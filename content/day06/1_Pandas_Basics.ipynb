## <p style="align: right; direction: rtl; float: right;">יצירת DataFrame ראשון</p>

<p style="text-align: right; direction: rtl;">
<strong>DataFrame</strong> זה כמו טבלת Excel - יש לו שורות ועמודות.
נתחיל ביצירה פשוטה מתוך מילון:
</p>

```python
import pandas as pd

# יצירת DataFrame מתוך מילון
# כל מפתח במילון יהפוך לשם עמודה
נתונים = {
    'שם': ['דני', 'שרה', 'יוסי', 'מיכל'],
    'גיל': [25, 30, 28, 35],
    'עיר': ['תל אביב', 'חיפה', 'ירושלים', 'באר שבע'],
    'משכורת': [10000, 15000, 12000, 18000]
}

df = pd.DataFrame(נתונים)
print(df)
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
יצרנו מילון Python רגיל, שבו כל מפתח מייצג עמודה וכל רשימה מכילה את הערכים באותה עמודה.<br>
Pandas אוטומטית יצר אינדקס (מספור השורות) - 0, 1, 2, 3.<br>
התוצאה היא טבלה מסודרת ונוחה לקריאה!
</p>

## <p style="align: right; direction: rtl; float: right;">קריאת נתונים מ-CSV</p>

<p style="text-align: right; direction: rtl;">
<strong>CSV</strong> (Comma Separated Values) זה פורמט קובץ נפוץ מאוד לשמירת טבלאות.<br>
Excel יכול לפתוח CSV, ורוב המערכות יודעות לייצא אליו נתונים.<br>
Pandas הופך קריאת CSV לפעולה של שורה אחת!
</p>

```python
# קריאת קובץ CSV
# הנתיב הוא יחסי למיקום המחברת
df_עובדים = pd.read_csv('../../resources/employees.csv')

# הצגת 5 שורות ראשונות
print("5 שורות ראשונות:")
print(df_עובדים.head())

print("\n" + "="*50 + "\n")

# הצגת 3 שורות אחרונות
print("3 שורות אחרונות:")
print(df_עובדים.tail(3))
```

<p style="text-align: right; direction: rtl;">
<strong>למה להשתמש ב-head() ו-tail()?</strong><br>
כשיש לנו מאות או אלפי שורות, אנחנו לא רוצים להדפיס את הכל!<br>
<code>head()</code> מציג את תחילת הטבלה - מעולה לבדוק שהקובץ נטען נכון.<br>
<code>tail()</code> מציג את סוף הטבלה - שימושי לוודא שכל הנתונים נטענו.
</p>

## <p style="align: right; direction: rtl; float: right;">מידע על ה-DataFrame</p>

<p style="text-align: right; direction: rtl;">
לפני שמתחילים לעבוד עם נתונים, חשוב להבין מה יש לנו בטבלה.<br>
Pandas מספק כלים נהדרים לכך:
</p>

```python
# מידע כללי על הטבלה
print("מידע על הטבלה:")
print(df_עובדים.info())

print("\n" + "="*50 + "\n")

# גודל הטבלה (שורות, עמודות)
print(f"גודל הטבלה: {df_עובדים.shape}")
print(f"מספר שורות: {df_עובדים.shape[0]}")
print(f"מספר עמודות: {df_עובדים.shape[1]}")

print("\n" + "="*50 + "\n")

# שמות העמודות
print("שמות העמודות:")
print(df_עובדים.columns.tolist())
```

<p style="text-align: right; direction: rtl;">
<strong>מה למדנו?</strong><br>
<code>info()</code> - מציג את סוג הנתונים בכל עמודה, האם יש ערכים חסרים, וכמה זיכרון הטבלה תופסת.<br>
<code>shape</code> - נותן טאפל (tuple) עם מספר השורות והעמודות.<br>
<code>columns</code> - נותן גישה לשמות כל העמודות.
</p>

## <p style="align: right; direction: rtl; float: right;">סטטיסטיקות בסיסיות</p>

<p style="text-align: right; direction: rtl;">
אחת הפונקציות החזקות של Pandas היא <code>describe()</code>.<br>
היא מחשבת אוטומטית סטטיסטיקות עבור כל העמודות המספריות!
</p>

```python
# סטטיסטיקות מפורטות
print(df_עובדים.describe())
```

<p style="text-align: right; direction: rtl;">
<strong>מה רואים כאן?</strong>
</p>

<ul style="text-align: right; direction: rtl;">
    <li><strong>count</strong> - כמה ערכים יש (לא כולל חסרים)</li>
    <li><strong>mean</strong> - ממוצע</li>
    <li><strong>std</strong> - סטיית תקן (מידת הפיזור של הנתונים)</li>
    <li><strong>min/max</strong> - ערך מינימלי ומקסימלי</li>
    <li><strong>25%, 50%, 75%</strong> - אחוזונים (percentiles)</li>
</ul>

<p style="text-align: right; direction: rtl;">
מדד 50% זה החציון (median) - הערך האמצעי כשממיינים את כל הנתונים.
</p>

## <p style="align: right; direction: rtl; float: right;">גישה לעמודות</p>

<p style="text-align: right; direction: rtl;">
יש 2 דרכים עיקריות לגשת לעמודה:
</p>

```python
# דרך 1: סוגריים מרובעים (כמו מילון)
print("גישה דרך סוגריים:")
print(df['שם'])

print("\n" + "="*50 + "\n")

# דרך 2: נקודה (אם אין רווחים בשם העמודה)
# לא עובד עם שמות עמודות בעברית ברוב המקרים
# print(df.name)  # עובד רק באנגלית ללא רווחים
```

<p style="text-align: right; direction: rtl;">
<strong>איזו דרך עדיפה?</strong><br>
סוגריים מרובעים (<code>df['עמודה']</code>) - תמיד עובד, גם עם רווחים ועברית.<br>
נקודה (<code>df.column</code>) - נוח יותר לכתיבה, אבל מוגבל לאנגלית ללא רווחים.
</p>

## <p style="align: right; direction: rtl; float: right;">בחירת כמה עמודות</p>

<p style="text-align: right; direction: rtl;">
לפעמים נרצה לראות רק חלק מהעמודות. נעביר רשימה של שמות:
</p>

```python
# בחירת עמודות ספציפיות
עמודות_נבחרות = ['שם', 'עיר']
print(df[עמודות_נבחרות])
```

<p style="text-align: right; direction: rtl;">
<strong>למה זה שימושי?</strong><br>
כשיש 50 עמודות אבל אנחנו מתעניינים רק ב-3, קל יותר לראות את מה שחשוב.<br>
זה גם חוסך זיכרון אם נשמור את ה-DataFrame המצומצם.
</p>

## <p style="align: right; direction: rtl; float: right;">גישה לשורות - loc ו-iloc</p>

<p style="text-align: right; direction: rtl;">
ב-Pandas יש 2 דרכים עיקריות לגשת לשורות:
</p>

<ul style="text-align: right; direction: rtl;">
    <li><strong>loc</strong> - גישה לפי שם האינדקס (label-based)</li>
    <li><strong>iloc</strong> - גישה לפי מיקום (position-based, כמו רשימה רגילה)</li>
</ul>

```python
# iloc - גישה לפי מיקום (מתחיל מ-0)
print("השורה הראשונה (מיקום 0):")
print(df.iloc[0])

print("\n" + "="*50 + "\n")

# iloc עם טווח
print("שורות 1-3 (מיקומים 1,2,3):")
print(df.iloc[1:4])  # שימו לב: 4 לא כלול!

print("\n" + "="*50 + "\n")

# loc - גישה לפי שם (כרגע האינדקס הוא 0,1,2...)
print("שורה באינדקס 0:")
print(df.loc[0])
```

<p style="text-align: right; direction: rtl;">
<strong>מתי להשתמש במה?</strong><br>
<code>iloc</code> - כשאתם חושבים במונחים של "תן לי שורה מספר 5" או "שורות 10-20".<br>
<code>loc</code> - כשיש לכם אינדקס משמעותי (למשל תאריכים או קוד עובד) ואתם רוצים לגשת לפי הערך.
</p>

## <p style="align: right; direction: rtl; float: right;">שמירת DataFrame ל-CSV</p>

<p style="text-align: right; direction: rtl;">
אחרי שעבדנו על הנתונים, נרצה לשמור אותם.<br>
Pandas הופך גם את זה לפשוט:
</p>

```python
# שמירה ל-CSV
df.to_csv('output.csv', index=False, encoding='utf-8-sig')

print("הקובץ נשמר בהצלחה!")
```

<p style="text-align: right; direction: rtl;">
<strong>הסברים על הפרמטרים:</strong><br>
<code>index=False</code> - לא לשמור את עמודת האינדקס (הספרור 0,1,2...) כעמודה בקובץ.<br>
<code>encoding='utf-8-sig'</code> - קידוד שמבטיח שעברית תיראה נכון גם ב-Excel.
</p>
