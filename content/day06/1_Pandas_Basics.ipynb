## <p style="align: right; direction: rtl; float: right;">מה זה Pandas?</p>

<p style="text-align: right; direction: rtl; float: right;">
Pandas היא ספרייה לניתוח נתונים. היא מספקת DataFrame - מבנה נתונים רב עוצמה לעבודה עם טבלאות.
</p>

```python
import pandas as pd

# יצירת DataFrame
data = {
    'שם': ['דני', 'שרה', 'יוסי'],
    'גיל': [25, 30, 28],
    'עיר': ['תל אביב', 'חיפה', 'ירושלים']
}

df = pd.DataFrame(data)
print(df)
```
