## <p style="align: right; direction: rtl; float: right;">סינון נתונים</p>

<p style="text-align: right; direction: rtl;">
אחת הפעולות הנפוצות ביותר בניתוח נתונים היא <strong>סינון</strong>.<br>
אנחנו רוצים למצוא רק את השורות שעומדות בתנאי מסוים.
</p>

<p style="text-align: right; direction: rtl;">
<strong>דוגמאות לשאלות שסינון עונה עליהן:</strong>
</p>

<ul style="text-align: right; direction: rtl;">
    <li>מי כל העובדים שמשכורתם מעל 15,000?</li>
    <li>אילו לקוחות גרים בתל אביב?</li>
    <li>מהם כל המוצרים שנמכרו ביוני?</li>
</ul>

```python
import pandas as pd

# יצירת דוגמה
נתונים = {
    'שם': ['דני', 'שרה', 'יוסי', 'מיכל', 'רועי'],
    'גיל': [25, 30, 28, 35, 22],
    'עיר': ['תל אביב', 'חיפה', 'ירושלים', 'תל אביב', 'חיפה'],
    'משכורת': [10000, 15000, 12000, 18000, 9000]
}

df = pd.DataFrame(נתונים)
print("הטבלה המלאה:")
print(df)
```

### <p style="align: right; direction: rtl; float: right;">סינון פשוט</p>

<p style="text-align: right; direction: rtl;">
הדרך הבסיסית לסנן היא ליצור "מסכה" של True/False:
</p>

```python
# יצירת מסכה - מערך של True/False
מסכה = df['גיל'] > 25
print("המסכה:")
print(מסכה)

print("\n" + "="*50 + "\n")

# שימוש במסכה לסינון
עובדים_מעל_25 = df[מסכה]
print("עובדים מעל גיל 25:")
print(עובדים_מעל_25)
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
כשכתבנו <code>df['גיל'] > 25</code>, Pandas בדק לכל שורה האם הגיל גדול מ-25.<br>
קיבלנו מערך של True/False - True בשורות שעומדות בתנאי.<br>
כשהעברנו את המסכה הזו ב-<code>df[מסכה]</code>, קיבלנו רק את השורות עם True!
</p>

### <p style="align: right; direction: rtl; float: right;">דרך קצרה</p>

<p style="text-align: right; direction: rtl;">
בפועל, לא צריך משתנה נפרד למסכה. אפשר לכתוב הכל בשורה אחת:
</p>

```python
# סינון ישיר - הדרך הנפוצה ביותר
משכורת_גבוהה = df[df['משכורת'] > 12000]
print("עובדים עם משכורת מעל 12,000:")
print(משכורת_גבוהה)
```

### <p style="align: right; direction: rtl; float: right;">תנאים מרובים - AND</p>

<p style="text-align: right; direction: rtl;">
לפעמים נרצה שכמה תנאים יתקיימו <strong>יחד</strong>.<br>
נשתמש ב-<code>&</code> (AND) ונשים כל תנאי בסוגריים:
</p>

```python
# סינון עם 2 תנאים (AND)
# חייבים סוגריים סביב כל תנאי!
תוצאה = df[(df['גיל'] > 25) & (df['עיר'] == 'תל אביב')]
print("עובדים מעל 25 שגרים בתל אביב:")
print(תוצאה)
```

<p style="text-align: right; direction: rtl;">
<strong>שימו לב!</strong> חובה להשתמש ב-<code>&</code> ולא ב-<code>and</code> של Python הרגיל.<br>
וחובה לשים סוגריים סביב כל תנאי! אחרת תקבלו שגיאה.
</p>

### <p style="align: right; direction: rtl; float: right;">תנאים מרובים - OR</p>

<p style="text-align: right; direction: rtl;">
אם נרצה ש<strong>לפחות אחד</strong> מהתנאים יתקיים, נשתמש ב-<code>|</code> (OR):
</p>

```python
# סינון עם OR - מספיק שתנאי אחד מתקיים
תוצאה = df[(df['עיר'] == 'חיפה') | (df['עיר'] == 'תל אביב')]
print("עובדים מחיפה או תל אביב:")
print(תוצאה)
```

### <p style="align: right; direction: rtl; float: right;">isin - בדיקת השתייכות לרשימה</p>

<p style="text-align: right; direction: rtl;">
כשיש לנו הרבה אפשרויות, <code>isin()</code> יותר נוח מ-OR רבים:
</p>

```python
# דרך יפה יותר עם isin
ערים = ['חיפה', 'תל אביב']
תוצאה = df[df['עיר'].isin(ערים)]
print("עובדים מהערים הנבחרות:")
print(תוצאה)
```

## <p style="align: right; direction: rtl; float: right;">מיון נתונים</p>

<p style="text-align: right; direction: rtl;">
לפעמים נרצה למיין את הטבלה לפי עמודה מסוימת.
</p>

```python
# מיון לפי גיל (עולה)
print("ממוין לפי גיל:")
print(df.sort_values('גיל'))

print("\n" + "="*50 + "\n")

# מיון לפי משכורת (יורד)
print("ממוין לפי משכורת (גבוה לנמוך):")
print(df.sort_values('משכורת', ascending=False))
```

<p style="text-align: right; direction: rtl;">
<strong>פרמטרים חשובים:</strong><br>
<code>ascending=True</code> (ברירת מחדל) - מיון עולה (קטן לגדול, A-Z).<br>
<code>ascending=False</code> - מיון יורד (גדול לקטן, Z-A).
</p>

## <p style="align: right; direction: rtl; float: right;">קיבוץ - groupby</p>

<p style="text-align: right; direction: rtl;">
<code>groupby</code> היא אחת הפונקציות החזקות ביותר ב-Pandas!<br>
היא מאפשרת לקבץ שורות לפי ערך משותף ולבצע חישובים על כל קבוצה.
</p>

<p style="text-align: right; direction: rtl;">
<strong>למשל:</strong>
</p>

<ul style="text-align: right; direction: rtl;">
    <li>מה הממוצע של משכורות לפי עיר?</li>
    <li>כמה עובדים יש בכל מחלקה?</li>
    <li>מה סכום המכירות לפי חודש?</li>
</ul>

```python
# ממוצע משכורת לפי עיר
print("ממוצע משכורת לפי עיר:")
print(df.groupby('עיר')['משכורת'].mean())

print("\n" + "="*50 + "\n")

# ספירה - כמה עובדים בכל עיר
print("מספר עובדים לפי עיר:")
print(df.groupby('עיר').size())
```

<p style="text-align: right; direction: rtl;">
<strong>איך זה עובד?</strong><br>
<code>groupby('עיר')</code> - מקבץ את כל השורות שיש להן אותה עיר.<br>
<code>['משכורת']</code> - בוחר את עמודת המשכורת.<br>
<code>.mean()</code> - מחשב ממוצע לכל קבוצה.
</p>

### <p style="align: right; direction: rtl; float: right;">פונקציות אגרגציה נפוצות</p>

<p style="text-align: right; direction: rtl;">
אחרי <code>groupby</code> אפשר להשתמש בפונקציות רבות:
</p>

```python
# מינימום, מקסימום, סכום
print("משכורת מקסימלית לפי עיר:")
print(df.groupby('עיר')['משכורת'].max())

print("\n" + "="*50 + "\n")

print("סכום משכורות לפי עיר:")
print(df.groupby('עיר')['משכורת'].sum())
```

### <p style="align: right; direction: rtl; float: right;">agg - כמה חישובים ביחד</p>

<p style="text-align: right; direction: rtl;">
אם רוצים כמה סטטיסטיקות בבת אחת, משתמשים ב-<code>agg()</code>:
</p>

```python
# כמה חישובים ביחד
print(df.groupby('עיר')['משכורת'].agg(['mean', 'min', 'max', 'count']))
```

## <p style="align: right; direction: rtl; float: right;">מיזוג טבלאות - merge</p>

<p style="text-align: right; direction: rtl;">
הרבה פעמים הנתונים שלנו מפוזרים בכמה טבלאות.<br>
<code>merge</code> מאפשר לחבר טבלאות לפי עמודה משותפת - כמו JOIN ב-SQL.
</p>

```python
# טבלה 1 - פרטי עובדים
עובדים = pd.DataFrame({
    'id': [1, 2, 3],
    'שם': ['דני', 'שרה', 'יוסי'],
    'מחלקה_id': [10, 20, 10]
})

# טבלה 2 - פרטי מחלקות
מחלקות = pd.DataFrame({
    'id': [10, 20, 30],
    'שם_מחלקה': ['פיתוח', 'שיווק', 'מכירות']
})

print("טבלת עובדים:")
print(עובדים)
print("\n" + "="*50 + "\n")
print("טבלת מחלקות:")
print(מחלקות)
```

```python
# מיזוג - חיבור העובדים למחלקות שלהם
תוצאה = pd.merge(
    עובדים,
    מחלקות,
    left_on='מחלקה_id',  # העמודה בטבלה השמאלית
    right_on='id',        # העמודה בטבלה הימנית
    how='left'            # סוג המיזוג
)

print("\nאחרי מיזוג:")
print(תוצאה)
```

<p style="text-align: right; direction: rtl;">
<strong>סוגי מיזוג (how):</strong>
</p>

<ul style="text-align: right; direction: rtl;">
    <li><strong>left</strong> - שומר את כל השורות מהטבלה השמאלית</li>
    <li><strong>right</strong> - שומר את כל השורות מהטבלה הימנית</li>
    <li><strong>inner</strong> (ברירת מחדל) - רק שורות שיש להן התאמה בשתי הטבלאות</li>
    <li><strong>outer</strong> - כל השורות משתי הטבלאות</li>
</ul>

## <p style="align: right; direction: rtl; float: right;">טיפול בערכים חסרים</p>

<p style="text-align: right; direction: rtl;">
בעולם האמיתי, לפעמים יש ערכים חסרים בנתונים.<br>
Pandas מסמן אותם כ-<code>NaN</code> (Not a Number).
</p>

```python
import numpy as np

# יצירת DataFrame עם ערכים חסרים
נתונים_חסרים = pd.DataFrame({
    'שם': ['דני', 'שרה', 'יוסי'],
    'גיל': [25, np.nan, 28],
    'עיר': ['תל אביב', 'חיפה', None]
})

print("נתונים עם ערכים חסרים:")
print(נתונים_חסרים)

print("\n" + "="*50 + "\n")

# בדיקה איפה יש ערכים חסרים
print("האם יש ערכים חסרים?")
print(נתונים_חסרים.isnull())
```

```python
# מילוי ערכים חסרים
ממולא = נתונים_חסרים.fillna({
    'גיל': נתונים_חסרים['גיל'].mean(),  # ממוצע
    'עיר': 'לא ידוע'
})

print("אחרי מילוי:")
print(ממולא)

print("\n" + "="*50 + "\n")

# מחיקת שורות עם ערכים חסרים
נקי = נתונים_חסרים.dropna()
print("אחרי מחיקת שורות חסרות:")
print(נקי)
```
