### <p style="align: right; direction: rtl; float: right;">פתרון תרגיל 1</p>

```python
import pandas as pd

# יצירת DataFrame
מוצרים = pd.DataFrame({
    'שם_מוצר': ['מחשב נייד', 'עכבר', 'מקלדת', 'מסך', 'אוזניות'],
    'מחיר': [3000, 50, 150, 800, 200],
    'מלאי': [5, 20, 15, 8, 12],
    'קטגוריה': ['מחשבים', 'אביזרים', 'אביזרים', 'מחשבים', 'אביזרים']
})

# 1. הדפסת ה-DataFrame
print("טבלת המוצרים:")
print(מוצרים)

print("\n" + "="*50 + "\n")

# 2. סטטיסטיקות
print("סטטיסטיקות:")
print(מוצרים.describe())

print("\n" + "="*50 + "\n")

# 3. המוצר היקר ביותר
מוצר_יקר = מוצרים.loc[מוצרים['מחיר'].idxmax()]
print(f"המוצר היקר ביותר: {מוצר_יקר['שם_מוצר']} - ₪{מוצר_יקר['מחיר']}")

print("\n" + "="*50 + "\n")

# 4. שווי מלאי
מוצרים['שווי_מלאי'] = מוצרים['מחיר'] * מוצרים['מלאי']
print("שווי מלאי לכל מוצר:")
print(מוצרים[['שם_מוצר', 'שווי_מלאי']])
print(f"\nשווי מלאי כולל: ₪{מוצרים['שווי_מלאי'].sum():,}")
```

### <p style="align: right; direction: rtl; float: right;">פתרון תרגיל 2</p>

```python
# 1. מוצרים מעל 100 ש"ח
print("מוצרים מעל ₪100:")
print(מוצרים[מוצרים['מחיר'] > 100])

print("\n" + "="*50 + "\n")

# 2. מוצרים עם מלאי נמוך
print("מוצרים עם מלאי נמוך (<10):")
print(מוצרים[מוצרים['מלאי'] < 10])

print("\n" + "="*50 + "\n")

# 3. ממוצע מחיר לפי קטגוריה
print("ממוצע מחיר לפי קטגוריה:")
print(מוצרים.groupby('קטגוריה')['מחיר'].mean())

print("\n" + "="*50 + "\n")

# 4. סך מלאי לפי קטגוריה
print("סך מלאי לפי קטגוריה:")
print(מוצרים.groupby('קטגוריה')['מלאי'].sum())
```

<p style="text-align: right; direction: rtl;">
צרו את הטבלאות, מזגו אותן, וחשבו את סכום ההכנסות לכל הזמנה.

### <p style="align: right; direction: rtl; float: right;">פתרון תרגיל 3</p>

```python
# יצירת טבלת הזמנות
הזמנות = pd.DataFrame({
    'מספר_הזמנה': [1001, 1002, 1003, 1004],
    'תאריך': ['2024-01-15', '2024-01-16', '2024-01-16', '2024-01-17'],
    'מוצר_id': [101, 102, 101, 103],
    'כמות': [2, 1, 3, 5]
})

# יצירת טבלת מוצרים
מוצרים_למיזוג = pd.DataFrame({
    'id': [101, 102, 103],
    'שם': ['מקלדת', 'עכבר', 'אוזניות'],
    'מחיר': [150, 50, 200]
})

print("טבלת הזמנות:")
print(הזמנות)
print("\n" + "="*50 + "\n")
print("טבלת מוצרים:")
print(מוצרים_למיזוג)

# מיזוג
הזמנות_מלאות = pd.merge(
    הזמנות,
    מוצרים_למיזוג,
    left_on='מוצר_id',
    right_on='id',
    how='left'
)

# חישוב סכום
הזמנות_מלאות['סכום'] = הזמנות_מלאות['כמות'] * הזמנות_מלאות['מחיר']

print("\n" + "="*50 + "\n")
print("הזמנות אחרי מיזוג:")
print(הזמנות_מלאות[['מספר_הזמנה', 'תאריך', 'שם', 'כמות', 'מחיר', 'סכום']])

print("\n" + "="*50 + "\n")
print(f"סך כל ההכנסות: ₪{הזמנות_מלאות['סכום'].sum():,}")
```
