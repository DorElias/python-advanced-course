# <p style="align: right; direction: rtl; float: right;">SQL מתקדם - UPDATE, DELETE, JOIN</p>

## <p style="align: right; direction: rtl; float: right;">חלק 1: עדכון נתונים - UPDATE</p>

```python
import sqlite3

# חיבור למסד
conn = sqlite3.connect('company.db')
cursor = conn.cursor()

# עדכון שכר
cursor.execute('''
    UPDATE employees
    SET salary = salary * 1.1
    WHERE department = 'הנדסה'
''')

print(f"✅ עודכנו {cursor.rowcount} עובדים")

# שמירה
conn.commit()

# בדיקה
cursor.execute("SELECT name, salary FROM employees WHERE department = 'הנדסה'")
for row in cursor.fetchall():
    print(f"{row[0]}: ₪{row[1]:,.2f}")

conn.close()
```

## <p style="align: right; direction: rtl; float: right;">חלק 2: מחיקת נתונים - DELETE</p>

```python
conn = sqlite3.connect('company.db')
cursor = conn.cursor()

# מחיקה עם תנאי
cursor.execute('''
    DELETE FROM employees
    WHERE age > 65
''')

print(f"🗑️  נמחקו {cursor.rowcount} רשומות")
conn.commit()
conn.close()
```

## <p style="align: right; direction: rtl; float: right;">חלק 3: JOIN - חיבור טבלאות</p>

```python
conn = sqlite3.connect('school.db')
cursor = conn.cursor()

# יצירת טבלאות
cursor.execute('''
    CREATE TABLE IF NOT EXISTS students (
        id INTEGER PRIMARY KEY,
        name TEXT,
        class_id INTEGER
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS classes (
        id INTEGER PRIMARY KEY,
        class_name TEXT,
        teacher TEXT
    )
''')

# נתונים לדוגמה
cursor.execute("DELETE FROM students")
cursor.execute("DELETE FROM classes")

classes = [
    (1, 'כיתה א', 'גב\' כהן'),
    (2, 'כיתה ב', 'מר לוי'),
    (3, 'כיתה ג', 'גב\' מזרחי')
]

students = [
    (1, 'דני', 1),
    (2, 'שרה', 1),
    (3, 'יוסי', 2),
    (4, 'רחל', 2),
    (5, 'משה', 3)
]

cursor.executemany('INSERT INTO classes VALUES (?,?,?)', classes)
cursor.executemany('INSERT INTO students VALUES (?,?,?)', students)
conn.commit()

# JOIN - חיבור הטבלאות
cursor.execute('''
    SELECT students.name, classes.class_name, classes.teacher
    FROM students
    JOIN classes ON students.class_id = classes.id
''')

print("📚 תלמידים וכיתות:\n")
for row in cursor.fetchall():
    print(f"{row[0]} לומד ב{row[1]} אצל {row[2]}")

conn.close()
```

## <p style="align: right; direction: rtl; float: right;">חלק 4: GROUP BY - קיבוץ</p>

```python
conn = sqlite3.connect('company.db')
cursor = conn.cursor()

# ספירת עובדים לפי מחלקה
cursor.execute('''
    SELECT department, COUNT(*) as num_employees, AVG(salary) as avg_salary
    FROM employees
    GROUP BY department
''')

print("📊 סטטיסטיקות לפי מחלקה:\n")
for row in cursor.fetchall():
    print(f"{row[0]}: {row[1]} עובדים, שכר ממוצע ₪{row[2]:,.2f}")

conn.close()
```
