# <p style="align: right; direction: rtl; float: right;">יום 5: עבודה עם מסדי נתונים - SQLite</p>

<p style="text-align: right; direction: rtl; float: right;">
היום נלמד איך לעבוד עם מסדי נתונים בפייתון!<br>
מסד נתונים הוא כמו אקסל חכם - מקום לשמור מידע בצורה מסודרת ולחפש בו במהירות.<br>
נשתמש ב-SQLite - מסד נתונים פשוט ומובנה בפייתון, מושלם להתחיל.
</p>

## <p style="align: right; direction: rtl; float: right;">מה זה SQL?</p>

<p style="text-align: right; direction: rtl; float: right;">
SQL (Structured Query Language) היא שפה מיוחדת לעבודה עם מסדי נתונים.<br>
היא מאפשרת לנו:<br>
✅ ליצור טבלאות<br>
✅ להוסיף מידע<br>
✅ לחפש ולסנן מידע<br>
✅ לעדכן ולמחוק מידע
</p>

## <p style="align: right; direction: rtl; float: right;">התחברות למסד נתונים</p>

<p style="text-align: right; direction: rtl; float: right;">
בואו נתחבר למסד נתונים (או ניצור אחד חדש אם הוא לא קיים):
</p>

```python
import sqlite3

# יצירת חיבור למסד נתונים
# אם הקובץ לא קיים, הוא ייווצר אוטומטית
connection = sqlite3.connect('my_database.db')

# יצירת cursor - כלי לביצוע פקודות SQL
cursor = connection.cursor()

print("התחברנו בהצלחה למסד הנתונים! 🎉")
```

## <p style="align: right; direction: rtl; float: right;">יצירת טבלה - CREATE TABLE</p>

<p style="text-align: right; direction: rtl; float: right;">
בואו ניצור טבלה לשמירת פרטי תלמידים.<br>
כל תלמיד יהיה עם: מזהה, שם, גיל ועיר.
</p>

```python
# יצירת טבלת תלמידים
cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER,
    city TEXT
)
''')

# שמירת השינויים
connection.commit()

print("הטבלה נוצרה בהצלחה! ✅")
```

<p style="text-align: right; direction: rtl; float: right;">
הסבר על הפקודה:<br>
📌 <strong>CREATE TABLE</strong> - יוצר טבלה חדשה<br>
📌 <strong>IF NOT EXISTS</strong> - יוצר רק אם הטבלה לא קיימת (למנוע שגיאות)<br>
📌 <strong>INTEGER PRIMARY KEY AUTOINCREMENT</strong> - מזהה ייחודי שעולה אוטומטית<br>
📌 <strong>TEXT</strong> - טיפוס טקסט (כמו str בפייתון)<br>
📌 <strong>NOT NULL</strong> - שדה חובה (לא יכול להיות ריק)
</p>

## <p style="align: right; direction: rtl; float: right;">הוספת נתונים - INSERT</p>

<p style="text-align: right; direction: rtl; float: right;">
עכשיו בואו נוסיף תלמידים לטבלה:
</p>

```python
# הוספת תלמיד אחד
cursor.execute('''
INSERT INTO students (name, age, city)
VALUES ('דני כהן', 25, 'תל אביב')
''')

connection.commit()
print("תלמיד אחד נוסף! ✅")
```

```python
# הוספת מספר תלמידים בבת אחת
students_data = [
    ('שרה לוי', 23, 'חיפה'),
    ('יוסי כהן', 27, 'ירושלים'),
    ('רחל אברהם', 22, 'באר שבע'),
    ('משה דוד', 26, 'תל אביב')
]

cursor.executemany('''
INSERT INTO students (name, age, city)
VALUES (?, ?, ?)
''', students_data)

connection.commit()
print(f"נוספו {len(students_data)} תלמידים! 🎉")
```

<p style="text-align: right; direction: rtl; float: right;">
💡 <strong>טיפ:</strong> הסימן <code>?</code> משמש כ-placeholder (מציין מיקום) לערכים.<br>
זה בטוח יותר ומונע התקפות SQL Injection.
</p>

## <p style="align: right; direction: rtl; float: right;">שליפת נתונים - SELECT</p>

<p style="text-align: right; direction: rtl; float: right;">
עכשיו נלמד לשלוף ולהציג את המידע שהכנסנו:
</p>

```python
# שליפת כל התלמידים
cursor.execute('SELECT * FROM students')

# קבלת כל השורות
all_students = cursor.fetchall()

print("כל התלמידים:")
for student in all_students:
    print(f"ID: {student[0]}, שם: {student[1]}, גיל: {student[2]}, עיר: {student[3]}")
```

```python
# שליפת עמודות ספציפיות בלבד
cursor.execute('SELECT name, city FROM students')

results = cursor.fetchall()
print("\nשמות וערים:")
for name, city in results:
    print(f"{name} מ{city}")
```

```python
# שליפת תלמיד אחד בלבד
cursor.execute('SELECT * FROM students LIMIT 1')

first_student = cursor.fetchone()
print(f"\nהתלמיד הראשון: {first_student}")
```

## <p style="align: right; direction: rtl; float: right;">סינון נתונים - WHERE</p>

<p style="text-align: right; direction: rtl; float: right;">
WHERE מאפשר לנו לסנן את התוצאות לפי תנאים:<br>
</p>

```python
# תלמידים מתל אביב בלבד
cursor.execute('''
SELECT name, age FROM students
WHERE city = 'תל אביב'
''')

tel_aviv_students = cursor.fetchall()
print("תלמידים מתל אביב:")
for name, age in tel_aviv_students:
    print(f"{name}, גיל {age}")
```

```python
# תלמידים מעל גיל 25
cursor.execute('''
SELECT name, age FROM students
WHERE age > 25
''')

older_students = cursor.fetchall()
print("\nתלמידים מעל גיל 25:")
for name, age in older_students:
    print(f"{name}, גיל {age}")
```

```python
# שילוב של מספר תנאים
cursor.execute('''
SELECT name, age, city FROM students
WHERE age >= 25 AND city = 'תל אביב'
''')

filtered_students = cursor.fetchall()
print("\nתלמידים מתל אביב מעל גיל 25:")
for name, age, city in filtered_students:
    print(f"{name}, גיל {age}, {city}")
```

## <p style="align: right; direction: rtl; float: right;">אופרטורים ב-WHERE</p>

```python
# LIKE - חיפוש חלקי (כמו contains)
cursor.execute('''
SELECT name FROM students
WHERE name LIKE '%כהן%'
''')

cohens = cursor.fetchall()
print("תלמידים עם שם משפחה 'כהן':")
for (name,) in cohens:
    print(f"- {name}")
```

```python
# IN - בדיקה האם הערך נמצא ברשימה
cursor.execute('''
SELECT name, city FROM students
WHERE city IN ('תל אביב', 'חיפה')
''')

big_cities = cursor.fetchall()
print("\nתלמידים מערים גדולות:")
for name, city in big_cities:
    print(f"{name} - {city}")
```

```python
# BETWEEN - בדיקת טווח
cursor.execute('''
SELECT name, age FROM students
WHERE age BETWEEN 23 AND 26
''')

age_range = cursor.fetchall()
print("\nתלמידים בגילאים 23-26:")
for name, age in age_range:
    print(f"{name}, גיל {age}")
```

## <p style="align: right; direction: rtl; float: right;">מיון תוצאות - ORDER BY</p>

```python
# מיון לפי גיל (מהצעיר למבוגר)
cursor.execute('''
SELECT name, age FROM students
ORDER BY age
''')

sorted_by_age = cursor.fetchall()
print("תלמידים ממוינים לפי גיל:")
for name, age in sorted_by_age:
    print(f"{name}, גיל {age}")
```

```python
# מיון הפוך (מהמבוגר לצעיר)
cursor.execute('''
SELECT name, age FROM students
ORDER BY age DESC
''')

sorted_desc = cursor.fetchall()
print("\nתלמידים לפי גיל (מהמבוגר לצעיר):")
for name, age in sorted_desc:
    print(f"{name}, גיל {age}")
```

## <p style="align: right; direction: rtl; float: right;">פונקציות צבירה</p>

```python
# COUNT - ספירת שורות
cursor.execute('SELECT COUNT(*) FROM students')
total_students = cursor.fetchone()[0]
print(f"סה\"כ תלמידים: {total_students}")

# AVG - ממוצע
cursor.execute('SELECT AVG(age) FROM students')
avg_age = cursor.fetchone()[0]
print(f"גיל ממוצע: {avg_age:.1f}")

# MIN - מינימום
cursor.execute('SELECT MIN(age) FROM students')
min_age = cursor.fetchone()[0]
print(f"הגיל הצעיר ביותר: {min_age}")

# MAX - מקסימום
cursor.execute('SELECT MAX(age) FROM students')
max_age = cursor.fetchone()[0]
print(f"הגיל המבוגר ביותר: {max_age}")
```

## <p style="align: right; direction: rtl; float: right;">סגירת החיבור</p>

<p style="text-align: right; direction: rtl; float: right;">
חשוב לסגור את החיבור למסד הנתונים כשסיימנו:
</p>

```python
# סגירת החיבור
connection.close()
print("החיבור למסד הנתונים נסגר בהצלחה! 👋")
```

## <p style="align: right; direction: rtl; float: right;">דוגמה מלאה - מערכת ניהול ספרייה</p>

```python
# יצירת מסד נתונים לספרייה
import sqlite3

conn = sqlite3.connect('library.db')
cursor = conn.cursor()

# יצירת טבלת ספרים
cursor.execute('''
CREATE TABLE IF NOT EXISTS books (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    author TEXT NOT NULL,
    year INTEGER,
    available INTEGER DEFAULT 1
)
''')

# הוספת ספרים
books = [
    ('הקצר הזה על האריה', 'דוד גרוסמן', 2014, 1),
    ('תמול שלשום', 'ש"י עגנון', 1945, 1),
    ('גת', 'יהושע קנז', 2013, 0),
    ('יונתן הקטן', 'ליאה גולדברג', 1950, 1)
]

cursor.executemany('INSERT INTO books (title, author, year, available) VALUES (?, ?, ?, ?)', books)
conn.commit()

# חיפוש ספרים זמינים
print("ספרים זמינים להשאלה:")
cursor.execute('SELECT title, author FROM books WHERE available = 1')
for title, author in cursor.fetchall():
    print(f"📖 {title} - {author}")

# חיפוש ספרים לפי מחבר
print("\nספרים של דוד גרוסמן:")
cursor.execute('SELECT title, year FROM books WHERE author LIKE "%גרוסמן%"')
for title, year in cursor.fetchall():
    print(f"📚 {title} ({year})")

conn.close()
print("\n✅ הספרייה נוצרה בהצלחה!")
```
