## <p style="align: right; direction: rtl; float: right;">חלק 1: מה זה Class?</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
<strong>Class (מחלקה)</strong> היא תבנית ליצירת עצמים.<br>
חשבו על זה כמו "תבנית עוגיות" - המחלקה היא התבנית, והעצמים (instances) הם העוגיות שאנחנו יוצרים ממנה.
</p>

### <p style="align: right; direction: rtl; float: right;">דוגמה פשוטה: מחלקת Person</p>

```python
class Person:
    """מחלקה שמייצגת אדם"""
    
    def __init__(self, name, age):
        """פעולת אתחול - רצה בזמן יצירת העצם"""
        self.name = name
        self.age = age
    
    def introduce(self):
        """פעולה שמציגה את האדם"""
        return f"שלום, שמי {self.name} ואני בן {self.age}"
    
    def have_birthday(self):
        """פעולה שמוסיפה שנה לגיל"""
        self.age += 1
        return f"🎉 יום הולדת שמח! עכשיו אני בן {self.age}"

# יצירת עצמים (instances)
dani = Person("דני", 25)
sara = Person("שרה", 30)

# שימוש בפעולות
print(dani.introduce())
print(sara.introduce())
print()
print(dani.have_birthday())
```

## <p style="align: right; direction: rtl; float: right;">חלק 2: דוגמה מעשית - מערכת ניהול משימות</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
בואו נבנה מערכת ניהול משימות (TODO List) עם Classes.<br>
נתחיל עם מחלקה בסיסית למשימה אחת:
</p>

```python
class Task:
    """מחלקה שמייצגת משימה"""
    
    def __init__(self, title, description=""):
        self.title = title
        self.description = description
        self.completed = False  # ברירת מחדל - המשימה לא הושלמה
    
    def mark_complete(self):
        """מסמנת את המשימה כהושלמה"""
        self.completed = True
        return f"✅ המשימה '{self.title}' הושלמה!"
    
    def mark_incomplete(self):
        """מסמנת את המשימה כלא הושלמה"""
        self.completed = False
        return f"⏳ המשימה '{self.title}' חזרה למצב לא הושלם"
    
    def __str__(self):
        """פעולת קסם להדפסה יפה"""
        status = "✅" if self.completed else "⏳"
        result = f"{status} {self.title}"
        if self.description:
            result += f"\n   📝 {self.description}"
        return result

# נסו את המחלקה
task1 = Task("לקנות חלב", "בדרך הביתה מהעבודה")
task2 = Task("להתקשר לדני")

print(task1)
print()
print(task2)
print()
print(task1.mark_complete())
print()
print(task1)
```

### <p style="align: right; direction: rtl; float: right;">עכשיו - מחלקת TodoList לניהול כל המשימות</p>

```python
class TodoList:
    """מחלקה שמנהלת רשימת משימות"""
    
    def __init__(self, name="רשימת המשימות שלי"):
        self.name = name
        self.tasks = []  # רשימה ריקה של משימות
    
    def add_task(self, task):
        """מוסיפה משימה לרשימה"""
        self.tasks.append(task)
        return f"➕ המשימה '{task.title}' נוספה"
    
    def remove_task(self, task_index):
        """מוחקת משימה לפי אינדקס"""
        if 0 <= task_index < len(self.tasks):
            removed = self.tasks.pop(task_index)
            return f"🗑️  המשימה '{removed.title}' נמחקה"
        else:
            return "❌ אינדקס לא תקין"
    
    def get_pending_tasks(self):
        """מחזירה רק משימות שעדיין לא הושלמו"""
        return [task for task in self.tasks if not task.completed]
    
    def get_completed_tasks(self):
        """מחזירה רק משימות שהושלמו"""
        return [task for task in self.tasks if task.completed]
    
    def show_summary(self):
        """מציגה סיכום של הרשימה"""
        total = len(self.tasks)
        completed = len(self.get_completed_tasks())
        pending = len(self.get_pending_tasks())
        
        print(f"\n📋 {self.name}")
        print("=" * 50)
        print(f"סה\"כ משימות: {total}")
        print(f"✅ הושלמו: {completed}")
        print(f"⏳ ממתינות: {pending}")
    
    def show_all_tasks(self):
        """מציגה את כל המשימות"""
        if not self.tasks:
            print("\n📪 אין משימות ברשימה")
            return
        
        print(f"\n📋 {self.name}")
        print("=" * 50)
        for i, task in enumerate(self.tasks):
            print(f"\n{i+1}. {task}")

# נסו את המערכת!
my_todo = TodoList("משימות השבוע")

# הוספת משימות
task1 = Task("לקנות חלב", "בדרך הביתה")
task2 = Task("לשלם חשבונות", "חשבון חשמל + מים")
task3 = Task("להתקשר לדני")
task4 = Task("לסיים פרויקט", "עד יום חמישי")

print(my_todo.add_task(task1))
print(my_todo.add_task(task2))
print(my_todo.add_task(task3))
print(my_todo.add_task(task4))

# הצגת כל המשימות
my_todo.show_all_tasks()

# סימון משימות כהושלמו
print("\n" + task1.mark_complete())
print(task3.mark_complete())

# הצגת סיכום
my_todo.show_summary()

# הצגה מחדש
my_todo.show_all_tasks()
```

## <p style="align: right; direction: rtl; float: right;">חלק 3: ירושה (Inheritance)</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
<strong>ירושה</strong> מאפשרת לנו ליצור מחלקה חדשה שיורשת תכונות ופעולות ממחלקה קיימת.<br>
זה שימושי כשיש לנו מחלקה כללית, ואנחנו רוצים ליצור גרסה מתמחה שלה.
</p>

### <p style="align: right; direction: rtl; float: right;">דוגמה: משימות דחופות</p>

```python
class UrgentTask(Task):
    """מחלקה למשימה דחופה - יורשת מ-Task"""
    
    def __init__(self, title, description="", deadline=""):
        # קריאה ל-__init__ של המחלקה האם
        super().__init__(title, description)
        self.deadline = deadline
    
    def __str__(self):
        """פעולת הדפסה משופרת למשימה דחופה"""
        status = "✅" if self.completed else "🚨"
        result = f"{status} {self.title} [דחוף!]"
        if self.description:
            result += f"\n   📝 {self.description}"
        if self.deadline:
            result += f"\n   ⏰ דדליין: {self.deadline}"
        return result

# נסו את המשימה הדחופה
urgent = UrgentTask(
    "להגיש דוח", 
    "דוח רבעוני למנכ\"ל",
    "יום חמישי 17:00"
)

print(urgent)
print()
print(urgent.mark_complete())
print()
print(urgent)
```

### <p style="align: right; direction: rtl; float: right;">דוגמה נוספת: משימות חוזרות</p>

```python
class RecurringTask(Task):
    """מחלקה למשימה חוזרת - יורשת מ-Task"""
    
    def __init__(self, title, description="", frequency="יומי"):
        super().__init__(title, description)
        self.frequency = frequency
        self.streak = 0  # כמה פעמים בזה אחר זה הושלמה
    
    def mark_complete(self):
        """סימון השלמה + עדכון streak"""
        self.completed = True
        self.streak += 1
        return f"✅ המשימה '{self.title}' הושלמה! רצף: {self.streak} 🔥"
    
    def reset(self):
        """איפוס המשימה ליום הבא"""
        self.completed = False
        return f"🔄 המשימה '{self.title}' אופסה ליום הבא"
    
    def break_streak(self):
        """שבירת הרצף (לא הושלמה)"""
        self.streak = 0
        self.completed = False
        return f"💔 הרצף נשבר. נתחיל מחדש!"
    
    def __str__(self):
        status = "✅" if self.completed else "🔁"
        result = f"{status} {self.title} ({self.frequency})"
        if self.description:
            result += f"\n   📝 {self.description}"
        if self.streak > 0:
            result += f"\n   🔥 רצף: {self.streak} ימים"
        return result

# נסו משימה חוזרת
exercise = RecurringTask(
    "להתאמן",
    "30 דקות ריצה",
    "יומי"
)

print(exercise)
print()

# סימולציה של 5 ימים
for day in range(1, 6):
    print(f"\nיום {day}:")
    print(exercise.mark_complete())
    print(exercise)
    print(exercise.reset())
```

## <p style="align: right; direction: rtl; float: right;">חלק 4: דוגמה מורכבת - מערכת ניהול עובדים</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
בואו נבנה מערכת ניהול עובדים עם ירושה מרובת רמות:
</p>

```python
class Employee:
    """מחלקת אב - עובד כללי"""
    
    def __init__(self, name, employee_id, salary):
        self.name = name
        self.employee_id = employee_id
        self.salary = salary
    
    def get_annual_bonus(self):
        """בונוס שנתי - 10% משכר"""
        return self.salary * 0.1
    
    def __str__(self):
        return f"👤 {self.name} (#{self.employee_id}) - ₪{self.salary:,}"


class Manager(Employee):
    """מנהל - יורש מעובד"""
    
    def __init__(self, name, employee_id, salary, team_size):
        super().__init__(name, employee_id, salary)
        self.team_size = team_size
    
    def get_annual_bonus(self):
        """בונוס מנהל - 20% + 1% לכל עובד בצוות"""
        base_bonus = self.salary * 0.2
        team_bonus = self.salary * 0.01 * self.team_size
        return base_bonus + team_bonus
    
    def __str__(self):
        return f"👔 {self.name} (#{self.employee_id}) - ₪{self.salary:,} [מנהל צוות {self.team_size}]"


class Developer(Employee):
    """מפתח - יורש מעובד"""
    
    def __init__(self, name, employee_id, salary, programming_languages):
        super().__init__(name, employee_id, salary)
        self.programming_languages = programming_languages
    
    def get_annual_bonus(self):
        """בונוס מפתח - 15% + 2% לכל שפת תכנות"""
        base_bonus = self.salary * 0.15
        lang_bonus = self.salary * 0.02 * len(self.programming_languages)
        return base_bonus + lang_bonus
    
    def __str__(self):
        langs = ", ".join(self.programming_languages)
        return f"💻 {self.name} (#{self.employee_id}) - ₪{self.salary:,} [{langs}]"


class SeniorDeveloper(Developer):
    """מפתח בכיר - יורש ממפתח"""
    
    def __init__(self, name, employee_id, salary, programming_languages, years_experience):
        super().__init__(name, employee_id, salary, programming_languages)
        self.years_experience = years_experience
    
    def get_annual_bonus(self):
        """בונוס מפתח בכיר - בונוס מפתח רגיל + 3% לכל שנת ניסיון"""
        base_bonus = super().get_annual_bonus()
        experience_bonus = self.salary * 0.03 * self.years_experience
        return base_bonus + experience_bonus
    
    def __str__(self):
        langs = ", ".join(self.programming_languages)
        return f"🌟 {self.name} (#{self.employee_id}) - ₪{self.salary:,} [{langs}] ({self.years_experience} שנות ניסיון)"


# נסו את המערכת!
employees = [
    Employee("דני כהן", 1001, 10000),
    Manager("שרה לוי", 2001, 15000, 5),
    Developer("יוסי מזרחי", 3001, 12000, ["Python", "JavaScript"]),
    SeniorDeveloper("רחל אברהם", 3002, 18000, ["Python", "Go", "Rust"], 8)
]

print("\n🏢 רשימת עובדים:")
print("=" * 70)
for emp in employees:
    bonus = emp.get_annual_bonus()
    print(f"\n{emp}")
    print(f"   💰 בונוס שנתי: ₪{bonus:,.2f}")

# חישוב סך הבונוסים
total_bonuses = sum(emp.get_annual_bonus() for emp in employees)
print(f"\n{'='*70}")
print(f"סה\"כ בונוסים שנתיים: ₪{total_bonuses:,.2f}")
```
