## <p style="align: right; direction: rtl; float: right;">תרגיל 3: פעולות קסם - ShoppingCart</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">

</p>

<ol style="text-align: right; direction: rtl; float: right; clear: both;">

## <p style="align: right; direction: rtl; float: right;">תרגיל 7: בונוס אתגר 🏆 - מערכת בנקאות</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
בנו מערכת בנקאות מלאה עם:
</p>

###  <p style="align: right; direction: rtl; float: right;">1. Account (מחלקת אב)</p>
<ul style="text-align: right; direction: rtl; float: right; clear: both;">
    <li>תכונות: account_number, owner_name, balance</li>
    <li><code>deposit(amount)</code> - הפקדה</li>
    <li><code>withdraw(amount)</code> - משיכה (לא יכול להיות שלילי)</li>
    <li><code>get_balance()</code> - יתרה נוכחית</li>
    <li>Property <code>balance</code> - לא ניתן לשנות ישירות (read-only)</li>
</ul>

### <p style="align: right; direction: rtl; float: right;">2. SavingsAccount (חשבון חיסכון - יורש מ-Account)</p>
<ul style="text-align: right; direction: rtl; float: right; clear: both;">
    <li>תכונה נוספת: interest_rate (אחוז ריבית שנתי)</li>
    <li><code>apply_interest()</code> - מוסיף ריבית ליתרה</li>
    <li><code>withdraw</code> - override: מוגבל ל-3 משיכות בחודש</li>
</ul>

### <p style="align: right; direction: rtl; float: right;">3. CheckingAccount (חשבון עו"ש - יורש מ-Account)</p>
<ul style="text-align: right; direction: rtl; float: right; clear: both;">
    <li>תכונה נוספת: overdraft_limit (מסגרת משיכת יתר)</li>
    <li><code>withdraw</code> - override: מאפשר משיכת יתר עד המסגרת</li>
    <li>Property <code>available_balance</code> - יתרה + מסגרת</li>
</ul>
