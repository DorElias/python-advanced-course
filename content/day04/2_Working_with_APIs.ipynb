## <p style="align: right; direction: rtl; float: right;">מה זה API?</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
<strong>API</strong> זה קיצור של Application Programming Interface.<br>
במילים פשוטות: זה "תפריט" שמאפשר לתוכנות לדבר אחת עם השנייה.<br><br>

דוגמה מהחיים:<br>
🔹 אתם הולכים למסעדה (האפליקציה שלכם)<br>
🔹 המלצר מביא תפריט (ה-API)<br>
🔹 אתם מזמינים ממנו (שולחים בקשה)<br>
🔹 המטבח מכין (השרת עובד)<br>
🔹 המלצר מביא לכם את האוכל (התשובה)<br><br>

אתם לא צריכים לדעת איך המטבח עובד - רק מה לבקש איך!
</p>

## <p style="align: right; direction: rtl; float: right;">מה זה JSON?</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
<strong>JSON</strong> (JavaScript Object Notation) זה פורמט לאחסון והעברת מידע.<br>
הוא נראה מאוד דומה למילונים ורשימות בפייתון!<br><br>

דוגמה ל-JSON:
</p>

```python
# זה JSON (כתוב כמחרוזת):
json_example = '''
{
    "name": "דני",
    "age": 30,
    "city": "תל אביב",
    "hobbies": ["ספורט", "קריאה", "בישול"]
}
'''

print("JSON לדוגמה:")
print(json_example)
```

<p style="text-align: right; direction: rtl; float: right; clear: both;">
שימו לב שזה דומה למילון פייתון, אבל יש הבדלים קטנים:<br>
🔹 ב-JSON משתמשים ב-<code>"</code> (גרשיים כפולים) תמיד<br>
🔹 אין <code>True/False</code> אלא <code>true/false</code> (אותיות קטנות)<br>
🔹 JSON זה מחרוזת, צריך להמיר אותו למילון פייתון<br>
</p>

## <p style="align: right; direction: rtl; float: right;">המרת JSON למילון פייתון</p>

```python
import json

json_string = '{"name": "שרה", "age": 25, "city": "חיפה"}'

# המרה ל-dictionary של פייתון
person = json.loads(json_string)

print("סוג המשתנה:", type(person))
print("התוכן:", person)
print()
print("גישה לערכים:")
print(f"שם: {person['name']}")
print(f"גיל: {person['age']}")
print(f"עיר: {person['city']}")
```

## <p style="align: right; direction: rtl; float: right;">קריאה ל-API עם תשובת JSON</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
בואו נשתמש ב-API של ציטוטים מפורסמים (quotable.io):<br>
זה API חינמי שנותן לנו ציטוטים אקראיים מאנשים מפורסמים.
</p>

```python
import requests

# נבקש ציטוט אקראי
url = 'https://api.quotable.io/random'
response = requests.get(url)

print("סטטוס:", response.status_code)
print()
print("התשובה כטקסט:")
print(response.text)
print()
print("סוג התשובה:", type(response.text))
```

<p style="text-align: right; direction: rtl; float: right; clear: both;">
אוקיי, קיבלנו JSON אבל זה עדיין טקסט. בואו נהפוך אותו למילון:
</p>

```python
# הדרך הארוכה (עם json.loads)
response = requests.get('https://api.quotable.io/random')
data = json.loads(response.text)

print("הציטוט:")
print(f'"{data["content"]}"')
print(f"- {data['author']}")
```

```python
# הדרך הקצרה והנוחה (מומלץ!)
response = requests.get('https://api.quotable.io/random')
data = response.json()  # 👈 requests עושה את ההמרה בשבילנו!

print("ציטוט אקראי:")
print(f'"{data["content"]}"')
print(f"- {data['author']}")
print()
print(f"אורך: {data['length']} תווים")
```

## <p style="align: right; direction: rtl; float: right;">API עם פרמטרים - תמונות של כלבים! 🐕</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
ה-API של dog.ceo נותן לנו תמונות אקראיות של כלבים.<br>
אפשר גם לבקש גזעים ספציפיים!
</p>

```python
# תמונה אקראית של כלב
url = 'https://dog.ceo/api/breeds/image/random'
response = requests.get(url)
data = response.json()

if response.status_code == 200:
    print("סטטוס הבקשה:", data['status'])
    print("לינק לתמונה:")
    print(data['message'])
    print()
    print("העתיקו את הלינק לדפדפן כדי לראות את הכלב! 🐕")
```

```python
# קבלת מספר תמונות של כלבים
url = 'https://dog.ceo/api/breeds/image/random/3'
response = requests.get(url)
data = response.json()

print("3 תמונות אקראיות של כלבים:")
print("=" * 50)
for i, image_url in enumerate(data['message'], 1):
    print(f"{i}. {image_url}")
```

```python
# בואו נראה אילו גזעים יש
url = 'https://dog.ceo/api/breeds/list/all'
response = requests.get(url)
data = response.json()

breeds = list(data['message'].keys())
print(f"יש {len(breeds)} גזעים!")
print()
print("דוגמאות לגזעים:")
for breed in breeds[:10]:  # רק 10 הראשונים
    print(f"- {breed}")
```

## <p style="align: right; direction: rtl; float: right;">טיפול בשגיאות - חשוב מאוד!</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
לא תמיד הכל עובד חלק. אולי אין אינטרנט, אולי השרת נפל, אולי הכתובת שגויה.<br>
חשוב לטפל בשגיאות כדי שהתוכנה שלנו לא תקרוס! 🛡️
</p>

```python
# דוגמה 1: בדיקת סטטוס קוד
url = 'https://dog.ceo/api/breed/unicorn/images/random'  # גזע שלא קיים
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print("הכל עבד!")
    print(data['message'])
else:
    print(f"❌ שגיאה! קוד סטטוס: {response.status_code}")
    print("התשובה מהשרת:", response.text)
```

```python
# דוגמה 2: טיפול בשגיאות רשת (אין אינטרנט, השרת לא עונה)
try:
    url = 'https://api.quotable.io/random'
    response = requests.get(url, timeout=5)  # חכה מקסימום 5 שניות
    
    if response.status_code == 200:
        data = response.json()
        print(f'"{data["content"]}"')
        print(f"- {data['author']}")
    else:
        print(f"השרת החזיר שגיאה: {response.status_code}")
        
except requests.exceptions.Timeout:
    print("❌ הבקשה לקחה יותר מדי זמן (timeout)")
except requests.exceptions.ConnectionError:
    print("❌ לא הצלחתי להתחבר לשרת - בדקו את חיבור האינטרנט")
except Exception as e:
    print(f"❌ קרתה שגיאה לא צפויה: {e}")
```

## <p style="align: right; direction: rtl; float: right;">פונקציה נוחה לבקשות API</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
בואו נכתוב פונקציה שמטפלת בכל השגיאות בשבילנו:
</p>

```python
def get_api_data(url, params=None, timeout=10):
    """
    פונקציה לשליחת בקשת GET ל-API עם טיפול בשגיאות
    
    Parameters:
    url (str): כתובת ה-API
    params (dict): פרמטרים לבקשה (אופציונלי)
    timeout (int): זמן המתנה מקסימלי בשניות
    
    Returns:
    dict: המידע מה-API או None במקרה של שגיאה
    """
    try:
        response = requests.get(url, params=params, timeout=timeout)
        
        if response.status_code == 200:
            return response.json()
        else:
            print(f"❌ קוד שגיאה: {response.status_code}")
            return None
            
    except requests.exceptions.Timeout:
        print("❌ הבקשה לקחה יותר מדי זמן")
        return None
    except requests.exceptions.ConnectionError:
        print("❌ בעיה בחיבור לאינטרנט")
        return None
    except Exception as e:
        print(f"❌ שגיאה: {e}")
        return None

# נסיון של הפונקציה
print("דוגמה 1: ציטוט")
data = get_api_data('https://api.quotable.io/random')
if data:
    print(f'"{data["content"]}" - {data["author"]}')

print()
print("דוגמה 2: כלב")
data = get_api_data('https://dog.ceo/api/breeds/image/random')
if data:
    print(f"תמונה: {data['message']}")
```

## <p style="align: right; direction: rtl; float: right;">דוגמה מורכבת: יוצרים אוסף ציטוטים</p>

```python
def collect_quotes(num_quotes=5):
    """
    אוספת מספר ציטוטים מ-API
    """
    quotes = []
    url = 'https://api.quotable.io/random'
    
    print(f"אוסף {num_quotes} ציטוטים...")
    print("=" * 60)
    
    for i in range(num_quotes):
        data = get_api_data(url)
        if data:
            quote_info = {
                'text': data['content'],
                'author': data['author'],
                'length': data['length']
            }
            quotes.append(quote_info)
            print(f"\n{i+1}. \"{quote_info['text']}\"")
            print(f"   - {quote_info['author']} ({quote_info['length']} תווים)")
    
    return quotes

# נאסוף 3 ציטוטים
my_quotes = collect_quotes(3)
print("\n" + "=" * 60)
print(f"סה\"כ נאספו {len(my_quotes)} ציטוטים")
```

## <p style="align: right; direction: rtl; float: right;">טיפים חשובים לעבודה עם APIs</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
📌 <strong>קראו את התיעוד</strong> - כל API יש לו תיעוד שמסביר איך להשתמש בו<br>
📌 <strong>בדקו limits</strong> - APIs חינמיים לרוב מגבילים כמה בקשות אפשר לשלוח<br>
📌 <strong>תמיד טפלו בשגיאות</strong> - השתמשו ב-try/except<br>
📌 <strong>בדקו את הסטטוס קוד</strong> - לפני שמשתמשים בתשובה<br>
📌 <strong>השתמשו ב-response.json()</strong> - במקום json.loads()<br>
📌 <strong>timeout חשוב</strong> - אל תחכו לנצח לתשובה<br>
</p>

## <p style="align: right; direction: rtl; float: right;">תרגיל: חקר API</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
בחרו אחד מה-APIs למעלה וכתבו פונקציה שמקבלת פרמטר ומחזירה מידע מעניין.
</p>

```python
# דוגמה: פונקציה שמחזירה תמונה של גזע כלב לפי בחירה
def get_dog_by_breed(breed):
    """
    מחזירה תמונה של כלב מגזע מסוים
    """
    url = f'https://dog.ceo/api/breed/{breed}/images/random'
    data = get_api_data(url)
    
    if data and data['status'] == 'success':
        return data['message']
    else:
        return None

# נסיון
image_url = get_dog_by_breed('corgi')
if image_url:
    print(f"תמונה של קורגי: {image_url}")
else:
    print("לא מצאתי את הגזע הזה")

# עכשיו התור שלכם! נסו ליצור פונקציה משלכם
```
