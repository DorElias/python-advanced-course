## <p style="align: right; direction: rtl; float: right;">הבקשה הראשונה שלנו!</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
בואו נשלח את הבקשה הראשונה שלנו - נבקש מידע על מזג האוויר בתל אביב:
</p>

```python
import requests

# שולחים בקשה לקבלת מזג אויר
response = requests.get('https://wttr.in/TelAviv?format=3')

# מה קיבלנו בחזרה?
print("מזג האוויר בתל אביב:")
print(response.text)
```

<p style="text-align: right; direction: rtl; float: right; clear: both;">
<strong>מה קרה כאן?</strong><br>
🔹 <code>requests.get()</code> - שלחנו בקשת GET ("תן לי מידע")<br>
🔹 העברנו כתובת של שירות מזג אויר<br>
🔹 <code>response.text</code> - קיבלנו את התשובה כטקסט<br>
</p>

## <p style="align: right; direction: rtl; float: right;">בדיקת סטטוס הבקשה</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
לא תמיד הבקשה מצליחה. חשוב לבדוק שהכל עבד כמו שצריך:<br>
🔹 קוד 200 = הכל בסדר ✅<br>
🔹 קוד 404 = לא מצאו את מה שביקשת ❌<br>
🔹 קוד 500 = שגיאה בשרת 🔥<br>
</p>

```python
response = requests.get('https://wttr.in/Jerusalem?format=3')

print(f"קוד סטטוס: {response.status_code}")

if response.status_code == 200:
    print("✅ הבקשה הצליחה!")
    print(f"מזג האוויר: {response.text}")
else:
    print(f"❌ משהו השתבש. קוד: {response.status_code}")
```

## <p style="align: right; direction: rtl; float: right;">פרמטרים בבקשה (Query Parameters)</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
לפעמים אנחנו רוצים לשלוח פרטים נוספים עם הבקשה.<br>
זה כמו למלא טופס - אנחנו מעבירים מידע שעוזר לשרת לתת לנו בדיוק מה שאנחנו רוצים.<br><br>

דוגמה: נבקש תחזית מפורטת יותר:
</p>

```python
# דרך 1: להוסיף את הפרמטרים בכתובת עצמה
response = requests.get('https://wttr.in/Haifa?format=3&lang=he')
print("דרך 1:")
print(response.text)
print()
```

```python
# דרך 2: להשתמש ב-params (מומלץ! יותר קריא ונוח)
url = 'https://wttr.in/Haifa'
params = {
    'format': '3',
    'lang': 'he'
}

response = requests.get(url, params=params)
print("דרך 2 (מומלצת):")
print(response.text)
```

<p style="text-align: right; direction: rtl; float: right; clear: both;">
<strong>למה דרך 2 יותר טובה?</strong><br>
✅ יותר קל לקרוא<br>
✅ קל להוסיף/להסיר פרמטרים<br>
✅ פייתון דואג לעצמו לקידוד נכון של תווים מיוחדים<br>
</p>

## <p style="align: right; direction: rtl; float: right;">דוגמה מעשית: בודק מזג אוויר לערים שונות</p>

```python
# רשימת ערים שאנחנו רוצים לבדוק
cities = ['TelAviv', 'Jerusalem', 'Haifa', 'Eilat', 'BeerSheva']

print("דו״ח מזג אוויר לערים בישראל")
print("=" * 40)

for city in cities:
    url = f'https://wttr.in/{city}'
    params = {'format': '3'}
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        print(response.text)
    else:
        print(f"❌ לא הצלחתי לקבל מידע על {city}")
    
    print("-" * 40)
```

## <p style="align: right; direction: rtl; float: right;">Headers - כותרות הבקשה</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
לפעמים השרת רוצה לדעת מי שולח את הבקשה, או איזה סוג מידע אנחנו רוצים לקבל.<br>
Headers זה כמו "מעטפה" שעליה כתוב פרטים על השולח והמכתב.
</p>

```python
url = 'https://wttr.in/TelAviv?format=3'

# נוסיף Headers שמזהים אותנו
headers = {
    'User-Agent': 'Python Course - Weather App',
    'Accept-Language': 'he-IL'
}

response = requests.get(url, headers=headers)
print(response.text)
```
