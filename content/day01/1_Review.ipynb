## <p style="align: right; direction: rtl; float: right;">חזרה: טיפוסי משתנים</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
בואו נזכור את טיפוסי המשתנים הבסיסיים שיש לנו בפייתון:
</p>

```python
# מספרים (int, float)
age = 25
height = 1.75

# מחרוזות (str)
name = "דני"
city = "תל אביב"

# בוליאניים (bool)
is_student = True
has_car = False

print(f"{name} בן {age}, גובה {height}מ'")
print(f"סטודנט: {is_student}, יש רכב: {has_car}")
```

### <p style="align: right; direction: rtl; float: right;">רשימות (Lists)</p>

```python
# רשימה של שמות
students = ["דני", "שרה", "יוסי", "רחל"]

# הוספה לרשימה
students.append("משה")

# מחיקה מרשימה
students.remove("יוסי")

# גישה לפי אינדקס
first_student = students[0]

# אורך הרשימה
num_students = len(students)

print(f"יש {num_students} תלמידים")
print(f"התלמיד הראשון: {first_student}")
print(f"כל התלמידים: {students}")
```

### <p style="align: right; direction: rtl; float: right;">מילונים (Dictionaries)</p>

```python
# מילון עם פרטי אדם
person = {
    "name": "דני כהן",
    "age": 28,
    "city": "תל אביב",
    "job": "מהנדס תוכנה"
}

# גישה לערכים
print(f"שם: {person['name']}")
print(f"גיל: {person['age']}")

# הוספת מפתח חדש
person["email"] = "dani@example.com"

# מעבר על כל המפתחות והערכים
for key, value in person.items():
    print(f"{key}: {value}")
```

## <p style="align: right; direction: rtl; float: right;">חזרה: תנאים ולולאות</p>

### <p style="align: right; direction: rtl; float: right;">תנאים (if/elif/else)</p>

```python
grade = 85

if grade >= 90:
    result = "מצוין"
elif grade >= 80:
    result = "טוב מאוד"
elif grade >= 70:
    result = "טוב"
else:
    result = "נכשל"

print(f"ציון {grade} = {result}")
```

### <p style="align: right; direction: rtl; float: right;">לולאות (for/while)</p>

```python
# לולאת for על רשימה
fruits = ["תפוח", "בננה", "תפוז", "אבטיח"]

print("פירות שיש לנו:")
for fruit in fruits:
    print(f"- {fruit}")
```

```python
# לולאת while
counter = 1

while counter <= 5:
    print(f"מספר: {counter}")
    counter += 1
```

## <p style="align: right; direction: rtl; float: right;">חזרה: פונקציות</p>

```python
def calculate_average(numbers):
    """מחשבת את הממוצע של רשימת מספרים"""
    if not numbers:  # בדיקה אם הרשימה ריקה
        return 0
    
    total = sum(numbers)
    count = len(numbers)
    average = total / count
    
    return average

# שימוש בפונקציה
grades = [85, 90, 78, 92, 88]
avg = calculate_average(grades)
print(f"הממוצע: {avg:.2f}")
```

### <p style="align: right; direction: rtl; float: right;">פונקציה עם פרמטרים אופציונליים</p>

```python
def greet(name, greeting="שלום"):
    """מברכת אדם עם ברכה מותאמת אישית"""
    return f"{greeting}, {name}!"

# קריאה עם הפרמטר ברירת המחדל
print(greet("דני"))

# קריאה עם ברכה מותאמת
print(greet("שרה", "בוקר טוב"))
```

## <p style="align: right; direction: rtl; float: right;">חזרה: עבודה עם קבצים</p>

### <p style="align: right; direction: rtl; float: right;">כתיבה לקובץ</p>

```python
# כתיבת רשימת משימות לקובץ
tasks = [
    "לקנות חלב",
    "לשלם חשבונות",
    "להתקשר לדני",
    "לסיים פרויקט"
]

with open("tasks.txt", "w", encoding="utf-8") as file:
    for task in tasks:
        file.write(task + "\n")

print("הקובץ נכתב בהצלחה!")
```

### <p style="align: right; direction: rtl; float: right;">קריאה מקובץ</p>

```python
# קריאת המשימות מהקובץ
with open("tasks.txt", "r", encoding="utf-8") as file:
    content = file.read()
    print("המשימות שלי:")
    print(content)
```

### <p style="align: right; direction: rtl; float: right;">קריאה שורה אחר שורה</p>

```python
# קריאת כל שורה בנפרד (שימושי לקבצים גדולים)
with open("tasks.txt", "r", encoding="utf-8") as file:
    for line_number, line in enumerate(file, 1):
        # strip() מסיר רווחים ושורות חדשות בהתחלה ובסוף
        line = line.strip()
        print(f"{line_number}. {line}")
```

## <p style="align: right; direction: rtl; float: right;">תרגיל משולב - מערכת ניהול אנשי קשר</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
בואו נשלב את כל מה שחזרנו ונבנה מערכת פשוטה לניהול אנשי קשר.<br>
המערכת תכלול:
</p>

<ul style="text-align: right; direction: rtl; float: right; clear: both;">
    <li>רשימת אנשי קשר (מילון)</li>
    <li>פונקציה להוספת איש קשר</li>
    <li>פונקציה לחיפוש איש קשר</li>
    <li>פונקציה להצגת כל אנשי הקשר</li>
    <li>שמירה וטעינה מקובץ</li>
</ul>

```python
# מערכת ניהול אנשי קשר

def add_contact(contacts, name, phone, email):
    """מוסיפה איש קשר חדש"""
    contacts[name] = {
        "phone": phone,
        "email": email
    }
    print(f"✅ {name} נוסף לרשימת אנשי הקשר")

def find_contact(contacts, name):
    """מחפשת איש קשר לפי שם"""
    if name in contacts:
        contact = contacts[name]
        print(f"📱 {name}:")
        print(f"   טלפון: {contact['phone']}")
        print(f"   אימייל: {contact['email']}")
    else:
        print(f"❌ {name} לא נמצא ברשימת אנשי הקשר")

def show_all_contacts(contacts):
    """מציגה את כל אנשי הקשר"""
    if not contacts:
        print("📪 אין אנשי קשר ברשימה")
        return
    
    print(f"\n📋 רשימת אנשי קשר ({len(contacts)} אנשים):")
    print("=" * 50)
    
    for name, info in contacts.items():
        print(f"\n👤 {name}")
        print(f"   📞 {info['phone']}")
        print(f"   📧 {info['email']}")

def save_contacts(contacts, filename):
    """שומרת את אנשי הקשר לקובץ"""
    with open(filename, "w", encoding="utf-8") as file:
        for name, info in contacts.items():
            line = f"{name}|{info['phone']}|{info['email']}\n"
            file.write(line)
    print(f"💾 אנשי הקשר נשמרו בקובץ {filename}")

def load_contacts(filename):
    """טוענת אנשי קשר מקובץ"""
    contacts = {}
    
    try:
        with open(filename, "r", encoding="utf-8") as file:
            for line in file:
                line = line.strip()
                if line:  # בדיקה שהשורה לא ריקה
                    name, phone, email = line.split("|")
                    contacts[name] = {
                        "phone": phone,
                        "email": email
                    }
        print(f"📂 {len(contacts)} אנשי קשר נטענו מהקובץ {filename}")
    except FileNotFoundError:
        print(f"⚠️  הקובץ {filename} לא קיים, מתחילים עם רשימה ריקה")
    
    return contacts
```

### <p style="align: right; direction: rtl; float: right;">נסו את המערכת!</p>

```python
# יצירת רשימת אנשי קשר חדשה
my_contacts = {}

# הוספת אנשי קשר
add_contact(my_contacts, "דני כהן", "050-1234567", "dani@example.com")
add_contact(my_contacts, "שרה לוי", "052-9876543", "sara@example.com")
add_contact(my_contacts, "יוסי מזרחי", "054-5555555", "yossi@example.com")

# הצגת כל אנשי הקשר
show_all_contacts(my_contacts)
```

```python
# חיפוש איש קשר
find_contact(my_contacts, "דני כהן")
find_contact(my_contacts, "רחל")
```

```python
# שמירה לקובץ
save_contacts(my_contacts, "contacts.txt")
```

```python
# טעינה מקובץ
loaded_contacts = load_contacts("contacts.txt")
show_all_contacts(loaded_contacts)
```
