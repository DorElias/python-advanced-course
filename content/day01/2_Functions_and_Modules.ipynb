### <p style="align: right; direction: rtl; float: right;">פרמטרים אופציונליים וערכי ברירת מחדל</p>

```python
def create_user(name, age, city="תל אביב", active=True):
    """יוצרת משתמש עם ערכי ברירת מחדל"""
    user = {
        "name": name,
        "age": age,
        "city": city,
        "active": active
    }
    return user

# דרכים שונות לקרוא לפונקציה
user1 = create_user("דני", 25)  # משתמש ברירות מחדל
user2 = create_user("שרה", 30, "חיפה")  # עוקף את city
user3 = create_user("יוסי", 28, active=False)  # עוקף רק את active
user4 = create_user(name="רחל", age=35, city="ירושלים", active=True)  # כל הפרמטרים בשמות

print(user1)
print(user2)
print(user3)
print(user4)
```

### <p style="align: right; direction: rtl; float: right;">*args - מספר משתנה של ארגומנטים</p>

```python
def sum_all(*numbers):
    """מחשבת סכום של כמה מספרים שרוצים"""
    total = 0
    for num in numbers:
        total += num
    return total

# אפשר לקרוא עם כמה מספרים שרוצים
print(sum_all(1, 2, 3))  # 6
print(sum_all(10, 20, 30, 40, 50))  # 150
print(sum_all(5))  # 5

def print_names(*names):
    """מדפיסה רשימת שמות"""
    print(f"יש לנו {len(names)} שמות:")
    for i, name in enumerate(names, 1):
        print(f"{i}. {name}")

print_names("דני", "שרה", "יוסי", "רחל", "משה")
```

### <p style="align: right; direction: rtl; float: right;">**kwargs - מילון של ארגומנטים בשמות</p>

```python
def create_product(**details):
    """יוצרת מוצר עם כל הפרטים שרוצים"""
    print("פרטי המוצר:")
    for key, value in details.items():
        print(f"  {key}: {value}")
    return details

# אפשר לשלוח כל פרמטר בשם שרוצים
product1 = create_product(name="מחשב", price=3500, brand="Dell", warranty=3)
print()
product2 = create_product(name="עכבר", price=50, color="שחור")

def build_query(table, **conditions):
    """בונה שאילתת SQL דינמית"""
    query = f"SELECT * FROM {table}"
    
    if conditions:
        where_parts = [f"{key} = '{value}'" for key, value in conditions.items()]
        query += " WHERE " + " AND ".join(where_parts)
    
    return query

print()
print(build_query("users"))
print(build_query("users", city="תל אביב"))
print(build_query("products", category="אלקטרוניקה", price=1000))
```

### <p style="align: right; direction: rtl; float: right;">שילוב של *args ו-**kwargs</p>

```python
def flexible_function(required_param, *args, **kwargs):
    """פונקציה גמישה מאוד"""
    print(f"פרמטר חובה: {required_param}")
    
    if args:
        print(f"פרמטרים נוספים: {args}")
    
    if kwargs:
        print("פרמטרים בשמות:")
        for key, value in kwargs.items():
            print(f"  {key} = {value}")

flexible_function("חובה")
print()
flexible_function("חובה", 1, 2, 3)
print()
flexible_function("חובה", 1, 2, name="דני", age=25)
```

### <p style="align: right; direction: rtl; float: right;">פונקציות שמחזירות מספר ערכים</p>

```python
def calculate_stats(numbers):
    """מחשבת סטטיסטיקות על רשימת מספרים"""
    if not numbers:
        return None, None, None, None
    
    total = sum(numbers)
    average = total / len(numbers)
    minimum = min(numbers)
    maximum = max(numbers)
    
    return total, average, minimum, maximum

# קבלת כל הערכים
grades = [85, 92, 78, 95, 88, 76]
sum_val, avg, min_val, max_val = calculate_stats(grades)

print(f"סכום: {sum_val}")
print(f"ממוצע: {avg:.2f}")
print(f"מינימום: {min_val}")
print(f"מקסימום: {max_val}")

# אפשר גם לקבל רק חלק מהערכים
_, average, _, _ = calculate_stats(grades)
print(f"\nרק הממוצע: {average:.2f}")
```

### <p style="align: right; direction: rtl; float: right;">פונקציות כפרמטרים (Callbacks)</p>

```python
def apply_operation(numbers, operation):
    """מפעילה פעולה על כל מספר ברשימה"""
    results = []
    for num in numbers:
        results.append(operation(num))
    return results

def double(x):
    return x * 2

def square(x):
    return x ** 2

def to_string(x):
    return f"המספר {x}"

numbers = [1, 2, 3, 4, 5]

print("כפל ב-2:", apply_operation(numbers, double))
print("בריבוע:", apply_operation(numbers, square))
print("למחרוזת:", apply_operation(numbers, to_string))
```

### <p style="align: right; direction: rtl; float: right;">Lambda Functions - פונקציות אנונימיות</p>

```python
# במקום לכתוב פונקציה מלאה, אפשר להשתמש ב-lambda
numbers = [1, 2, 3, 4, 5]

# עם lambda
doubled = apply_operation(numbers, lambda x: x * 2)
print("כפל ב-2 עם lambda:", doubled)

# דוגמאות נוספות
students = [
    {"name": "דני", "grade": 85},
    {"name": "שרה", "grade": 92},
    {"name": "יוסי", "grade": 78},
    {"name": "רחל", "grade": 95}
]

# מיון לפי ציון
sorted_by_grade = sorted(students, key=lambda s: s["grade"])
print("\nמיון לפי ציון:")
for student in sorted_by_grade:
    print(f"  {student['name']}: {student['grade']}")

# מיון לפי שם
sorted_by_name = sorted(students, key=lambda s: s["name"])
print("\nמיון לפי שם:")
for student in sorted_by_name:
    print(f"  {student['name']}: {student['grade']}")
```

### <p style="align: right; direction: rtl; float: right;">מודולים שימושיים - datetime</p>

```python
from datetime import datetime, timedelta, date

# תאריך ושעה נוכחיים
now = datetime.now()
print(f"עכשיו: {now}")
print(f"שעה: {now.hour}:{now.minute:02d}")
print(f"תאריך: {now.day}/{now.month}/{now.year}")

# עיצוב מותאם אישית
formatted = now.strftime("%d/%m/%Y %H:%M:%S")
print(f"\nמעוצב: {formatted}")

# חישובי תאריכים
tomorrow = now + timedelta(days=1)
next_week = now + timedelta(weeks=1)
yesterday = now - timedelta(days=1)

print(f"\nמחר: {tomorrow.strftime('%d/%m/%Y')}")
print(f"שבוע הבא: {next_week.strftime('%d/%m/%Y')}")
print(f"אתמול: {yesterday.strftime('%d/%m/%Y')}")

# יצירת תאריך ספציפי
birthday = datetime(1995, 6, 15)
age_days = (now - birthday).days
age_years = age_days // 365
print(f"\nגיל בימים: {age_days}")
print(f"גיל בשנים: {age_years}")
```

### <p style="align: right; direction: rtl; float: right;">מודולים שימושיים - collections</p>

```python
from collections import Counter, defaultdict, namedtuple

# Counter - ספירת אלמנטים
words = "שלום עולם שלום פייתון עולם שלום".split()
word_count = Counter(words)
print("ספירת מילים:")
for word, count in word_count.items():
    print(f"  {word}: {count}")

print(f"\nהמילה הנפוצה ביותר: {word_count.most_common(1)}")

# defaultdict - מילון עם ערך ברירת מחדל
grades_by_student = defaultdict(list)
grades_by_student["דני"].append(85)
grades_by_student["דני"].append(90)
grades_by_student["שרה"].append(95)

print("\nציונים לפי תלמיד:")
for student, grades in grades_by_student.items():
    print(f"  {student}: {grades}")

# namedtuple - tuple עם שמות לשדות
Point = namedtuple('Point', ['x', 'y'])
p1 = Point(10, 20)
p2 = Point(x=5, y=15)

print(f"\nנקודה 1: x={p1.x}, y={p1.y}")
print(f"נקודה 2: x={p2.x}, y={p2.y}")
```

## <p style="align: right; direction: rtl; float: right;">תרגיל מעשי - מערכת ניהול ספרייה</p>

```python
from datetime import datetime, timedelta
from collections import defaultdict
import random

class Library:
    """מערכת ניהול ספרייה"""
    
    def __init__(self):
        self.books = []
        self.borrowed = defaultdict(list)  # מי שאל מה
        
    def add_book(self, title, author, copies=1):
        """הוספת ספר לספרייה"""
        book = {
            "id": len(self.books) + 1,
            "title": title,
            "author": author,
            "total_copies": copies,
            "available_copies": copies
        }
        self.books.append(book)
        return f"✅ הספר '{title}' נוסף ({copies} עותקים)"
    
    def borrow_book(self, user, book_id, days=14):
        """השאלת ספר"""
        book = next((b for b in self.books if b["id"] == book_id), None)
        
        if not book:
            return "❌ ספר לא נמצא"
        
        if book["available_copies"] == 0:
            return f"❌ אין עותקים זמינים של '{book['title']}'"
        
        book["available_copies"] -= 1
        due_date = datetime.now() + timedelta(days=days)
        
        self.borrowed[user].append({
            "book_id": book_id,
            "title": book["title"],
            "borrowed_date": datetime.now(),
            "due_date": due_date
        })
        
        return f"📚 {user} שאל את '{book['title']}' עד {due_date.strftime('%d/%m/%Y')}"
    
    def return_book(self, user, book_id):
        """החזרת ספר"""
        if user not in self.borrowed:
            return "❌ אין לך ספרים מושאלים"
        
        for i, item in enumerate(self.borrowed[user]):
            if item["book_id"] == book_id:
                book = next((b for b in self.books if b["id"] == book_id), None)
                book["available_copies"] += 1
                
                # בדיקת איחור
                days_late = (datetime.now() - item["due_date"]).days
                
                self.borrowed[user].pop(i)
                
                if days_late > 0:
                    fine = days_late * 2  # 2 שקל ליום
                    return f"⚠️ הספר הוחזר באיחור של {days_late} ימים. קנס: ₪{fine}"
                else:
                    return f"✅ תודה על החזרת '{item['title']}' בזמן!"
        
        return "❌ הספר לא נמצא ברשימת ההשאלות שלך"
    
    def show_available_books(self):
        """הצגת ספרים זמינים"""
        print("\n📚 ספרים זמינים:")
        print("=" * 60)
        for book in self.books:
            if book["available_copies"] > 0:
                print(f"#{book['id']}: {book['title']} - {book['author']}")
                print(f"      זמינים: {book['available_copies']}/{book['total_copies']}")
    
    def show_user_books(self, user):
        """הצגת ספרים מושאלים למשתמש"""
        if user not in self.borrowed or not self.borrowed[user]:
            print(f"\n{user} - אין ספרים מושאלים")
            return
        
        print(f"\n📖 ספרים מושאלים ל-{user}:")
        print("=" * 60)
        for item in self.borrowed[user]:
            days_left = (item["due_date"] - datetime.now()).days
            status = "⏰ דחוף!" if days_left < 3 else "✅"
            print(f"{status} {item['title']}")
            print(f"      החזרה עד: {item['due_date'].strftime('%d/%m/%Y')} ({days_left} ימים)")

# נסו את המערכת!
lib = Library()

# הוספת ספרים
print(lib.add_book("הארי פוטר", "ג.ק. רולינג", 3))
print(lib.add_book("1984", "ג'ורג' אורוול", 2))
print(lib.add_book("מלחמת הכוכבים", "ג'ורג' לוקאס", 1))

# השאלת ספרים
print("\n" + lib.borrow_book("דני", 1))
print(lib.borrow_book("שרה", 1))
print(lib.borrow_book("יוסי", 2))

# הצגת ספרים זמינים
lib.show_available_books()

# הצגת ספרים מושאלים
lib.show_user_books("דני")
lib.show_user_books("שרה")

# החזרת ספר
print("\n" + lib.return_book("דני", 1))
```
