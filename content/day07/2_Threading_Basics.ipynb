# <p style="align: right; direction: rtl; float: right;">Threading - הרצה מקבילית</p>

## <p style="align: right; direction: rtl; float: right;">מה זה Threading?</p>

<p style="text-align: right; direction: rtl;">
Threading מאפשר לתוכנית לעשות כמה דברים **במקביל** במקום בזה אחר זה.
</p>

<p style="text-align: right; direction: rtl;">
<strong>דוגמאות לשימוש:</strong>
</p>

<ul style="text-align: right; direction: rtl;">
    <li>הורדת מספר קבצים במקביל</li>
    <li>עיבוד נתונים בזמן שמציגים ממשק למשתמש</li>
    <li>שליחת מיילים למספר נמענים במקביל</li>
    <li>בדיקת מספר אתרים בו-זמנית</li>
</ul>

## <p style="align: right; direction: rtl; float: right;">דוגמה פשוטה</p>

```python
import threading
import time

def task(name, duration):
    print(f"משימה {name} התחילה")
    time.sleep(duration)
    print(f"משימה {name} הסתיימה")

# הרצה רגילה - זה אחר זה
print("=== ללא Threading ===")
start = time.time()
task("A", 2)
task("B", 2)
print(f"סה\"כ: {time.time() - start:.1f} שניות\n")

# עם Threading - במקביל!
print("=== עם Threading ===")
start = time.time()

t1 = threading.Thread(target=task, args=("A", 2))
t2 = threading.Thread(target=task, args=("B", 2))

t1.start()
t2.start()

t1.join()  # המתן לסיום
t2.join()

print(f"סה\"כ: {time.time() - start:.1f} שניות")
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה?</strong><br>
ללא Threading: 2 + 2 = 4 שניות<br>
עם Threading: שתי המשימות רצו במקביל = ~2 שניות!
</p>

## <p style="align: right; direction: rtl; float: right;">דוגמה מעשית - הורדת קבצים</p>

```python
import requests
import threading
import time

def download_image(url, filename):
    """הורדת תמונה"""
    try:
        response = requests.get(url, timeout=10)
        with open(filename, 'wb') as f:
            f.write(response.content)
        print(f"✅ {filename} הורד")
    except Exception as e:
        print(f"❌ שגיאה ב-{filename}: {e}")

# רשימת תמונות להורדה
urls = [
    ("https://dog.ceo/api/breeds/image/random", "dog1.jpg"),
    ("https://dog.ceo/api/breeds/image/random", "dog2.jpg"),
    ("https://dog.ceo/api/breeds/image/random", "dog3.jpg"),
]

# הורדה במקביל
threads = []
start = time.time()

for url, filename in urls:
    t = threading.Thread(target=download_image, args=(url, filename))
    threads.append(t)
    t.start()

# המתנה לכולם
for t in threads:
    t.join()

print(f"\n🎉 הכל הסתיים ב-{time.time() - start:.1f} שניות")
```
