## <p style="align: right; direction: rtl; float: right;">שליחת מייל HTML מעוצב</p>

<p style="text-align: right; direction: rtl;">
מיילים יכולים להיות יותר מסתם טקסט!<br>
אפשר להוסיף עיצוב, צבעים, קישורים ועוד.
</p>

```python
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# יצירת מייל
msg = MIMEMultipart('alternative')
msg['Subject'] = 'דוח שבועי - מערכת המכירות'
msg['From'] = 'system@company.com'
msg['To'] = 'manager@company.com'

# גרסת טקסט פשוט (לתמיכה במיילים שלא תומכים ב-HTML)
טקסט = '''
דוח שבועי
===========
סך מכירות: 45,000 ₪
מספר הזמנות: 23
'''

# גרסת HTML מעוצבת
html = '''
<html>
  <body dir="rtl">
    <h2 style="color: #2c3e50;">📊 דוח שבועי - מערכת המכירות</h2>
    <div style="background-color: #ecf0f1; padding: 20px; border-radius: 5px;">
      <p><strong>תקופה:</strong> 01/06/2024 - 07/06/2024</p>
      <table style="width: 100%; border-collapse: collapse;">
        <tr style="background-color: #3498db; color: white;">
          <th style="padding: 10px; text-align: right;">מדד</th>
          <th style="padding: 10px; text-align: right;">ערך</th>
        </tr>
        <tr>
          <td style="padding: 10px; border-bottom: 1px solid #bdc3c7;">סך מכירות</td>
          <td style="padding: 10px; border-bottom: 1px solid #bdc3c7; font-weight: bold; color: #27ae60;">45,000 ₪</td>
        </tr>
        <tr>
          <td style="padding: 10px; border-bottom: 1px solid #bdc3c7;">מספר הזמנות</td>
          <td style="padding: 10px; border-bottom: 1px solid #bdc3c7;">23</td>
        </tr>
        <tr>
          <td style="padding: 10px;">ממוצע להזמנה</td>
          <td style="padding: 10px;">1,956 ₪</td>
        </tr>
      </table>
    </div>
    <p style="color: #7f8c8d; font-size: 12px; margin-top: 20px;">
      מייל אוטומטי ממערכת הניהול | לא להשיב למייל זה
    </p>
  </body>
</html>
'''

# הוספת שתי הגרסאות
חלק_טקסט = MIMEText(טקסט, 'plain', 'utf-8')
חלק_html = MIMEText(html, 'html', 'utf-8')

msg.attach(חלק_טקסט)
msg.attach(חלק_html)

print("מייל HTML מוכן!")
print("\nכך יראה המייל:")
print("=" * 50)
print("📊 דוח שבועי - מערכת המכירות")
print("סך מכירות: 45,000 ₪")
print("מספר הזמנות: 23")
```

חלק מלקוחות המייל לא תומכים ב-HTML או חוסמים אותו.<br>
שליחת גם גרסת טקסט מבטיחה שכולם יוכלו לקרוא את המייל.

## <p style="align: right; direction: rtl; float: right;">שליחת מייל עם קובץ מצורף</p>

<p style="text-align: right; direction: rtl;">
אחת הפונקציות השימושיות ביותר - צירוף קבצים!
</p>

```python
from email.mime.base import MIMEBase
from email import encoders
import os

def שלח_מייל_עם_קובץ(נמען, נושא, תוכן, נתיב_קובץ):
    """
    פונקציה לשליחת מייל עם קובץ מצורף
    
    Args:
        נמען: כתובת המייל של הנמען
        נושא: נושא המייל
        תוכן: תוכן המייל
        נתיב_קובץ: נתיב מלא לקובץ שרוצים לצרף
    """
    msg = MIMEMultipart()
    msg['From'] = 'sender@example.com'
    msg['To'] = נמען
    msg['Subject'] = נושא
    
    # הוספת תוכן המייל
    msg.attach(MIMEText(תוכן, 'plain', 'utf-8'))
    
    # צירוף הקובץ
    if os.path.exists(נתיב_קובץ):
        שם_קובץ = os.path.basename(נתיב_קובץ)
        
        # קריאת הקובץ
        with open(נתיב_קובץ, 'rb') as קובץ:
            חלק = MIMEBase('application', 'octet-stream')
            חלק.set_payload(קובץ.read())
        
        # הצפנת הקובץ
        encoders.encode_base64(חלק)
        
        # הוספת header
        חלק.add_header(
            'Content-Disposition',
            f'attachment; filename= {שם_קובץ}'
        )
        
        msg.attach(חלק)
        print(f"הקובץ '{שם_קובץ}' צורף בהצלחה!")
    else:
        print(f"שגיאה: הקובץ {נתיב_קובץ} לא נמצא")
    
    return msg

# דוגמה לשימוש
msg = שלח_מייל_עם_קובץ(
    נמען='boss@company.com',
    נושא='דוח חודשי - יוני 2024',
    תוכן='שלום,\n\nמצורף דוח המכירות החודשי.\n\nבברכה',
    נתיב_קובץ='report_june.xlsx'
)

print("\nהמייל מוכן לשליחה (עם קובץ מצורף)")
```

## <p style="align: right; direction: rtl; float: right;">שליחה לכמה נמענים</p>

<p style="text-align: right; direction: rtl;">
לפעמים צריך לשלוח את אותו מייל לרשימת אנשים.
</p>

```python
def שלח_לרשימה(רשימת_נמענים, נושא, תוכן):
    """
    שולח מייל לרשימת נמענים
    
    Args:
        רשימת_נמענים: רשימה של כתובות מייל
        נושא: נושא המייל
        תוכן: תוכן המייל
    """
    # דרך 1: שליחה לכולם ביחד (כולם רואים את כל הכתובות)
    msg = MIMEMultipart()
    msg['From'] = 'sender@example.com'
    msg['To'] = ', '.join(רשימת_נמענים)  # מחבר עם פסיקים
    msg['Subject'] = נושא
    msg.attach(MIMEText(תוכן, 'plain', 'utf-8'))
    
    print(f"שליחה ל-{len(רשימת_נמענים)} נמענים במייל אחד")
    
    # דרך 2: BCC - שליחה נסתרת (אף אחד לא רואה את האחרים)
    msg2 = MIMEMultipart()
    msg2['From'] = 'sender@example.com'
    msg2['Bcc'] = ', '.join(רשימת_נמענים)
    msg2['Subject'] = נושא
    msg2.attach(MIMEText(תוכן, 'plain', 'utf-8'))
    
    print("\nאו: שליחה ב-BCC (נמענים לא רואים אחד את השני)")
    
    return msg, msg2

# דוגמה
נמענים = [
    'employee1@company.com',
    'employee2@company.com',
    'employee3@company.com'
]

msg, msg_bcc = שלח_לרשימה(
    נמענים,
    'עדכון חשוב מהנהלה',
    'שלום לכולם,\n\nרצינו לעדכן...'
)
```

<p style="text-align: right; direction: rtl;">
<strong>מתי להשתמש במה?</strong><br>
<code>To</code> - כשכולם צריכים לדעת מי עוד קיבל את המייל (דיון קבוצתי).<br>
<code>Cc</code> - "לידיעה" - אנשים שצריכים לדעת אבל לא חייבים לענות.<br>
<code>Bcc</code> - כשכל אחד צריך לקבל אבל לא לדעת מי עוד (ניוזלטרים, התראות).
</p>

## <p style="align: right; direction: rtl; float: right;">פונקציה מלאה - קלף פונקציות</p>

<p style="text-align: right; direction: rtl;">
הנה פונקציה מקיפה שמשלבת את הכל:
</p>

```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders
import os

class EmailSender:
    """מחלקה לשליחת מיילים בצורה נוחה"""
    
    def __init__(self, smtp_server, port, email, password):
        self.smtp_server = smtp_server
        self.port = port
        self.email = email
        self.password = password
    
    def send_email(self, to, subject, body, attachments=None, html=False):
        """
        שליחת מייל
        
        Args:
            to: נמען או רשימת נמענים
            subject: נושא
            body: תוכן המייל
            attachments: רשימת נתיבי קבצים לצירוף (אופציונלי)
            html: האם התוכן הוא HTML (False = טקסט רגיל)
        """
        # יצירת המייל
        msg = MIMEMultipart()
        msg['From'] = self.email
        msg['To'] = to if isinstance(to, str) else ', '.join(to)
        msg['Subject'] = subject
        
        # הוספת תוכן
        content_type = 'html' if html else 'plain'
        msg.attach(MIMEText(body, content_type, 'utf-8'))
        
        # צירוף קבצים
        if attachments:
            for file_path in attachments:
                if os.path.exists(file_path):
                    self._attach_file(msg, file_path)
        
        # שליחה
        try:
            server = smtplib.SMTP(self.smtp_server, self.port)
            server.starttls()
            server.login(self.email, self.password)
            server.send_message(msg)
            server.quit()
            return True, "המייל נשלח בהצלחה!"
        except Exception as e:
            return False, f"שגיאה בשליחה: {e}"
    
    def _attach_file(self, msg, file_path):
        """צירוף קובץ למייל"""
        filename = os.path.basename(file_path)
        with open(file_path, 'rb') as f:
            part = MIMEBase('application', 'octet-stream')
            part.set_payload(f.read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', f'attachment; filename= {filename}')
        msg.attach(part)

# דוגמה לשימוש
# sender = EmailSender(
#     smtp_server='smtp.gmail.com',
#     port=587,
#     email='your-email@gmail.com',
#     password='your-app-password'
# )

# הצלחה, הודעה = sender.send_email(
#     to='recipient@example.com',
#     subject='בדיקה',
#     body='זהו מייל בדיקה',
#     attachments=['report.pdf']
# )

print("המחלקה EmailSender מוכנה לשימוש!")
```
