### <p style="align: right; direction: rtl; float: right;">יצירת Workbook חדש</p>

````python
from openpyxl import Workbook

# יצירת workbook חדש
wb = Workbook()

# קבלת הגליון הפעיל (sheet)
ws = wb.active
ws.title = "הגליון הראשון"  # שינוי שם הגליון

# כתיבת נתונים
ws['A1'] = "שם"
ws['B1'] = "גיל"
ws['C1'] = "עיר"

ws['A2'] = "דני כהן"
ws['B2'] = 25
ws['C2'] = "תל אביב"

ws['A3'] = "שרה לוי"
ws['B3'] = 30
ws['C3'] = "חיפה"

# שמירה
wb.save('my_first_excel.xlsx')
print("✅ הקובץ my_first_excel.xlsx נוצר!")
````

### <p style="align: right; direction: rtl; float: right;">קריאת קובץ קיים</p>

````python
from openpyxl import load_workbook

# פתיחת קובץ קיים
wb = load_workbook('my_first_excel.xlsx')

# קבלת הגליון לפי שם
ws = wb['הגליון הראשון']

# קריאת ערכים
print("📊 נתונים מהקובץ:\n")
print(f"{ws['A1'].value:<15} {ws['B1'].value:<10} {ws['C1'].value}")
print("-" * 40)
print(f"{ws['A2'].value:<15} {ws['B2'].value:<10} {ws['C2'].value}")
print(f"{ws['A3'].value:<15} {ws['B3'].value:<10} {ws['C3'].value}")
````

## <p style="align: right; direction: rtl; float: right;">חלק 2: עבודה עם שורות ועמודות</p>

### <p style="align: right; direction: rtl; float: right;">כתיבה דרך append</p>

````python
wb = Workbook()
ws = wb.active
ws.title = "עובדים"

# כתיבת כותרות
ws.append(["שם", "מחלקה", "שכר", "תאריך תחילת עבודה"])

# כתיבת נתונים
employees = [
    ["דני כהן", "הנדסה", 15000, "2020-03-15"],
    ["שרה לוי", "שיווק", 12000, "2019-07-22"],
    ["יוסי מזרחי", "מכירות", 13500, "2021-01-10"],
    ["רחל אברהם", "הנדסה", 16000, "2018-11-05"],
]

for emp in employees:
    ws.append(emp)

wb.save('employees.xlsx')
print("✅ employees.xlsx נוצר עם 4 עובדים")
````

### <p style="align: right; direction: rtl; float: right;">קריאת כל השורות</p>

````python
wb = load_workbook('employees.xlsx')
ws = wb.active

print("👥 רשימת עובדים:\n")

# דרך 1: מעבר על כל השורות
for row in ws.iter_rows(min_row=2, values_only=True):  # min_row=2 מדלג על הכותרת
    name, dept, salary, hire_date = row
    print(f"👤 {name}")
    print(f"   מחלקה: {dept}")
    print(f"   שכר: ₪{salary:,}")
    print(f"   תחילת עבודה: {hire_date}")
    print()
````

### <p style="align: right; direction: rtl; float: right;">קריאה למילונים</p>

````python
wb = load_workbook('employees.xlsx')
ws = wb.active

# קבלת הכותרות
headers = [cell.value for cell in ws[1]]

# המרה לרשימת מילונים
employees = []
for row in ws.iter_rows(min_row=2, values_only=True):
    employee = dict(zip(headers, row))
    employees.append(employee)

print(f"טענו {len(employees)} עובדים\n")

# עכשיו אפשר לעבוד כמו עם CSV
high_earners = [e for e in employees if e['שכר'] > 14000]
print(f"עובדים עם שכר מעל 14,000:")
for emp in high_earners:
    print(f"  • {emp['שם']} - ₪{emp['שכר']:,}")
````

## <p style="align: right; direction: rtl; float: right;">חלק 3: עיצוב וסגנון</p>

````python
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side

wb = Workbook()
ws = wb.active
ws.title = "דוח מכירות"

# כותרת
ws['A1'] = "דוח מכירות חודשי"
ws['A1'].font = Font(size=16, bold=True, color="FFFFFF")
ws['A1'].fill = PatternFill(start_color="4472C4", end_color="4472C4", fill_type="solid")
ws['A1'].alignment = Alignment(horizontal="center", vertical="center")
ws.merge_cells('A1:D1')  # איחוד תאים

# כותרות עמודות
headers = ["מוצר", "כמות", "מחיר יח'", "סה\"כ"]
ws.append(headers)

# עיצוב כותרות
header_fill = PatternFill(start_color="D9E1F2", end_color="D9E1F2", fill_type="solid")
header_font = Font(bold=True)

for cell in ws[2]:
    cell.fill = header_fill
    cell.font = header_font
    cell.alignment = Alignment(horizontal="center")

# נתונים
sales_data = [
    ["מחשב נייד", 5, 3500, "=B3*C3"],
    ["עכבר", 20, 50, "=B4*C4"],
    ["מקלדת", 15, 120, "=B5*C5"],
    ["מסך", 8, 800, "=B6*C6"],
]

for row in sales_data:
    ws.append(row)

# שורת סיכום
ws.append(["סה\"כ", "=SUM(B3:B6)", "", "=SUM(D3:D6)"])

# עיצוב שורת סיכום
summary_row = ws[ws.max_row]
summary_fill = PatternFill(start_color="FFF2CC", end_color="FFF2CC", fill_type="solid")
summary_font = Font(bold=True)

for cell in summary_row:
    cell.fill = summary_fill
    cell.font = summary_font

# מסגרות
thin_border = Border(
    left=Side(style='thin'),
    right=Side(style='thin'),
    top=Side(style='thin'),
    bottom=Side(style='thin')
)

for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=1, max_col=4):
    for cell in row:
        cell.border = thin_border

# התאמת רוחב עמודות
ws.column_dimensions['A'].width = 20
ws.column_dimensions['B'].width = 10
ws.column_dimensions['C'].width = 12
ws.column_dimensions['D'].width = 12

wb.save('sales_report.xlsx')
print("✅ sales_report.xlsx נוצר עם עיצוב מלא!")
print("   פתחו את הקובץ ב-Excel כדי לראות את התוצאה")
````

## <p style="align: right; direction: rtl; float: right;">חלק 4: עבודה עם מספר גליונות</p>

````python
wb = Workbook()

# הסרת הגליון הדיפולטיבי
wb.remove(wb.active)

# יצירת גליון עובדים
ws_employees = wb.create_sheet("עובדים")
ws_employees.append(["שם", "מחלקה", "שכר"])
ws_employees.append(["דני", "הנדסה", 15000])
ws_employees.append(["שרה", "שיווק", 12000])

# יצירת גליון מחלקות
ws_departments = wb.create_sheet("מחלקות")
ws_departments.append(["מחלקה", "מנהל", "תקציב"])
ws_departments.append(["הנדסה", "יוסי כהן", 500000])
ws_departments.append(["שיווק", "מיכל לוי", 300000])

# יצירת גליון סיכום
ws_summary = wb.create_sheet("סיכום", 0)  # 0 = גליון ראשון
ws_summary['A1'] = "דוח משאבי אנוש"
ws_summary['A3'] = "סך עובדים:"
ws_summary['B3'] = "=COUNTA(עובדים!A2:A100)"  # ספירה אוטומטית
ws_summary['A4'] = "סך שכר:"
ws_summary['B4'] = "=SUM(עובדים!C2:C100)"

wb.save('company_data.xlsx')
print("✅ company_data.xlsx נוצר עם 3 גליונות!\n")

# קריאה חזרה
wb = load_workbook('company_data.xlsx')
print("📋 גליונות בקובץ:")
for sheet_name in wb.sheetnames:
    print(f"  • {sheet_name}")
````

## <p style="align: right; direction: rtl; float: right;">חלק 5: המרה בין CSV ל-Excel</p>

````python
import csv

# CSV -> Excel
def csv_to_excel(csv_file, excel_file, sheet_name="Data"):
    """המרת CSV ל-Excel"""
    wb = Workbook()
    ws = wb.active
    ws.title = sheet_name
    
    with open(csv_file, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            ws.append(row)
    
    wb.save(excel_file)
    return f"✅ {csv_file} -> {excel_file}"

# Excel -> CSV
def excel_to_csv(excel_file, csv_file, sheet_name=None):
    """המרת Excel ל-CSV"""
    wb = load_workbook(excel_file)
    ws = wb[sheet_name] if sheet_name else wb.active
    
    with open(csv_file, 'w', encoding='utf-8', newline='') as file:
        writer = csv.writer(file)
        for row in ws.iter_rows(values_only=True):
            writer.writerow(row)
    
    return f"✅ {excel_file} -> {csv_file}"

# נסו את ההמרות
print(csv_to_excel('../../resources/employees.csv', 'employees_from_csv.xlsx', 'עובדים'))
print(excel_to_csv('employees.xlsx', 'employees_from_excel.csv'))
````

## <p style="align: right; direction: rtl; float: right;">חלק 6: פרויקט מעשי - מחולל דוחות</p>

````python
from datetime import datetime
from openpyxl.chart import BarChart, Reference

class ReportGenerator:
    """מחולל דוחות Excel"""
    
    def __init__(self):
        self.wb = Workbook()
        self.wb.remove(self.wb.active)  # הסרת גליון ברירת מחדל
    
    def create_sales_report(self, sales_data):
        """יצירת דוח מכירות"""
        ws = self.wb.create_sheet("מכירות")
        
        # כותרת ראשית
        ws['A1'] = f"דוח מכירות - {datetime.now().strftime('%d/%m/%Y')}"
        ws['A1'].font = Font(size=14, bold=True)
        ws.merge_cells('A1:E1')
        
        # כותרות עמודות
        headers = ["מוצר", "קטגוריה", "כמות", "מחיר", "סה\"כ"]
        ws.append(headers)
        
        # עיצוב כותרות
        for cell in ws[2]:
            cell.font = Font(bold=True)
            cell.fill = PatternFill(start_color="4472C4", end_color="4472C4", fill_type="solid")
            cell.font = Font(bold=True, color="FFFFFF")
        
        # נתונים
        for sale in sales_data:
            total = sale['quantity'] * sale['price']
            ws.append([
                sale['product'],
                sale['category'],
                sale['quantity'],
                sale['price'],
                total
            ])
        
        # שורת סיכום
        last_row = ws.max_row
        ws.append([
            "סה\"כ",
            "",
            f"=SUM(C3:C{last_row})",
            "",
            f"=SUM(E3:E{last_row})"
        ])
        
        # עיצוב סיכום
        for cell in ws[last_row + 1]:
            cell.font = Font(bold=True)
            cell.fill = PatternFill(start_color="FFF2CC", end_color="FFF2CC", fill_type="solid")
        
        return "✅ גליון מכירות נוצר"
    
    def create_summary_sheet(self):
        """יצירת גליון סיכום"""
        ws = self.wb.create_sheet("סיכום", 0)
        
        ws['A1'] = "סיכום כללי"
        ws['A1'].font = Font(size=16, bold=True)
        
        ws['A3'] = "סך הכנסות:"
        ws['B3'] = "=מכירות!E" + str(self.wb['מכירות'].max_row)
        ws['B3'].font = Font(size=14, bold=True, color="008000")
        
        ws['A4'] = "מספר מוצרים:"
        ws['B4'] = f"=COUNTA(מכירות!A3:A{self.wb['מכירות'].max_row - 1})"
        
        ws['A5'] = "ממוצע מכירה:"
        ws['B5'] = f"=AVERAGE(מכירות!E3:E{self.wb['מכירות'].max_row - 1})"
        
        return "✅ גליון סיכום נוצר"
    
    def save(self, filename):
        """שמירת הדוח"""
        self.wb.save(filename)
        return f"💾 הדוח נשמר: {filename}"


# נסו את המערכת!
sales_data = [
    {'product': 'מחשב נייד', 'category': 'אלקטרוניקה', 'quantity': 5, 'price': 3500},
    {'product': 'עכבר', 'category': 'אלקטרוניקה', 'quantity': 20, 'price': 50},
    {'product': 'שולחן', 'category': 'ריהוט', 'quantity': 3, 'price': 800},
    {'product': 'כיסא', 'category': 'ריהוט', 'quantity': 10, 'price': 350},
    {'product': 'מסך', 'category': 'אלקטרוניקה', 'quantity': 8, 'price': 800},
]

report = ReportGenerator()
print(report.create_sales_report(sales_data))
print(report.create_summary_sheet())
print(report.save('monthly_report.xlsx'))

print("\n🎉 הדוח מוכן! פתחו את monthly_report.xlsx")
````
