# <p style="align: right; direction: rtl; float: right;">עבודה עם JSON</p>

## <p style="align: right; direction: rtl; float: right;">חלק 1: מה זה JSON?</p>

<p style="text-align: right; direction: rtl;">
<strong>JSON</strong> = JavaScript Object Notation<br>
פורמט קל משקל לאחסון והעברת נתונים. נראה כמעט זהה למילונים ורשימות בפייתון!
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה JSON כל כך חשוב?</strong>
</p>

<ul style="text-align: right; direction: rtl;">
    <li><strong>הפורמט המוביל באינטרנט</strong> - כל API מודרני מדבר JSON! (Facebook, Twitter, Google...)</li>
    <li><strong>שמירת הגדרות ותצורה</strong> - רוב האפליקציות שומרות הגדרות ב-JSON</li>
    <li><strong>מבנים מורכבים</strong> - תומך במילונים בתוך רשימות בתוך מילונים... עד אינסוף!</li>
    <li><strong>קריא לבני אדם וגם למחשבים</strong> - אפשר לפתוח ולהבין, אבל גם קל לתוכניות לעבד</li>
    <li><strong>קומפקטי</strong> - לוקח פחות מקום מ-XML ופורמטים אחרים</li>
</ul>

<p style="text-align: right; direction: rtl;">
<strong>מתי נשתמש ב-JSON?</strong><br>
• כשעובדים עם APIs (קבלת/שליחת נתונים לשרתים)<br>
• שמירת הגדרות אפליקציה<br>
• העברת נתונים בין מערכות שונות<br>
• כל פעם שצריך לשמור מבנה נתונים מורכב (לא רק טבלה פשוטה)
</p>

<p style="text-align: right; direction: rtl;">
<strong>ההבדל בין CSV ל-JSON:</strong><br>
• <strong>CSV</strong> טוב ל<strong>טבלאות</strong> - שורות ועמודות פשוטות<br>
• <strong>JSON</strong> טוב ל<strong>מבנים מורכבים</strong> - עצים, קינונים, היררכיות
</p>

## <p style="align: right; direction: rtl; float: right;">חלק 2: קריאת JSON</p>

<p style="text-align: right; direction: rtl;">
קריאת JSON בפייתון היא <strong>פשוטה להפליא</strong>!<br>
הסיבה: JSON דומה מאוד למבני הנתונים של פייתון (מילונים ורשימות).
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה כל כך נוח?</strong><br>
פייתון מבינה JSON "מהטבע" - קוראים קובץ JSON, ומיד מקבלים מילון פייתון!<br>
אין צורך בהמרות מסובכות או עיבוד מיוחד - פשוט <code>json.load()</code> ומקבלים את הנתונים.
</p>

```python
import json

# קריאת קובץ JSON
with open('../../resources/settings.json', 'r', encoding='utf-8') as file:
    settings = json.load(file)  # load = טעינה מקובץ, מחזיר מילון!

print("⚙️  הגדרות המערכת:\n")
print(f"שם האפליקציה: {settings['app_name']}")
print(f"גרסה: {settings['version']}")
print(f"שפה: {settings['language']}")
print(f"ערכת נושא: {settings['theme']}")

# גישה לנתונים מקוננים - מילון בתוך מילון!
print(f"\n📧 הגדרות אימייל:")
print(f"  שרת SMTP: {settings['email']['smtp_server']}")
print(f"  פורט: {settings['email']['smtp_port']}")
print(f"  TLS: {settings['email']['use_tls']}")
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
<code>json.load(file)</code> קרא את קובץ ה-JSON והמיר אותו <strong>אוטומטית</strong> למילון פייתון!<br>
עכשיו <code>settings</code> הוא פשוט מילון רגיל, ואפשר לגשת לערכים שלו ב-<code>settings['key']</code>.<br>
<strong>נתונים מקוננים:</strong> שימו לב שב-<code>settings['email']</code> יש עוד מילון - זה מקונן!<br>
אפשר לגשת לערכים פנימיים ב-<code>settings['email']['smtp_server']</code> - מילון בתוך מילון.<br>
<strong>חשוב:</strong> בדיוק כמו ב-CSV, חייבים <code>encoding='utf-8'</code> לעברית!
</p>

### <p style="align: right; direction: rtl; float: right;">עבודה עם נתונים מקוננים</p>

<p style="text-align: right; direction: rtl;">
אחד היתרונות הגדולים של JSON הוא היכולת <strong>לקנן</strong> מבני נתונים.<br>
במקום טבלה שטוחה כמו ב-CSV, אפשר ליצור היררכיות עמוקות ומורכבות.
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה שימושי?</strong><br>
דמיינו שאתם שומרים הגדרות של אפליקציה.<br>
יש לכם קטגוריות (אימייל, התראות, תצוגה), וכל קטגוריה יש לה הגדרות משלה.<br>
JSON מאפשר לקבץ הכל בצורה מאורגנת וברורה!
</p>

```python
# מעבר על כל התכונות וערכיהן
print("🔧 תכונות המערכת:\n")
for feature, enabled in settings['features'].items():
    status = "✅ פעיל" if enabled else "❌ כבוי"
    print(f"  {feature}: {status}")

# בדיקת ערכים מקוננים עם תנאים
if settings['notifications']['enabled']:
    print("\n🔔 התראות מופעלות")
    if settings['notifications']['sound']:
        print("   🔊 צליל: פעיל")
    if settings['notifications']['popup']:
        print("   💬 חלון קופץ: פעיל")
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
עברנו על <code>settings['features']</code> שהוא מילון בתוך המילון הראשי.<br>
כל מפתח (<code>feature</code>) מקבל ערך <code>True</code> או <code>False</code>, ואנחנו בודקים אותו ומציגים סטטוס.<br>
בחלק השני, ניגשנו לעומק של 2 רמות: <code>settings['notifications']['enabled']</code> - זה מקונן!<br>
<strong>למה זה חזק?</strong> במקום עשרות משתנים נפרדים, הכל מאורגן בהיררכיה ברורה.
</p>

## <p style="align: right; direction: rtl; float: right;">חלק 3: כתיבה ל-JSON</p>

<p style="text-align: right; direction: rtl;">
כתיבת JSON היא <strong>הפוכה מקריאה</strong> - מקבלים מבני נתונים של פייתון (מילונים, רשימות) והופכים אותם ל-JSON.
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה חשוב?</strong><br>
כשבונים אפליקציה, צריך לשמור הגדרות, מצב משתמש, או כל נתון מורכב.<br>
JSON הוא הדרך הטובה ביותר לעשות את זה - פשוט, קריא, וקל לטעון בחזרה.
</p>

```python
# יצירת נתונים חדשים - מבנה מורכב!
user_profile = {
    "name": "דני כהן",
    "email": "dani@example.com",
    "age": 28,
    "premium": True,
    "preferences": {  # מילון בתוך מילון!
        "language": "he",
        "theme": "dark",
        "notifications": True
    },
    "interests": ["תכנות", "מוזיקה", "ספורט"],  # רשימה!
    "stats": {  # עוד מילון מקונן
        "login_count": 45,
        "last_login": "2024-01-15",
        "projects": 12
    }
}

# שמירה לקובץ JSON
with open('user_profile.json', 'w', encoding='utf-8') as file:
    json.dump(user_profile, file, ensure_ascii=False, indent=2)
    # ensure_ascii=False: שומר עברית כמו שצריך (לא \uXXXX)
    # indent=2: מעצב יפה עם הזחות (קל לקריאה!)

print("✅ פרופיל המשתמש נשמר!\n")

# קריאה והצגה של מה ששמרנו
print("תוכן הקובץ:")
with open('user_profile.json', 'r', encoding='utf-8') as file:
    print(file.read())
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
יצרנו מילון פייתון מורכב עם מספר רמות של קינון: מילונים בתוך מילונים, ורשימות.<br>
<code>json.dump(user_profile, file, ...)</code> המיר את כל המבנה הזה לפורמט JSON ושמר אותו לקובץ!<br>
<strong>שני פרמטרים חשובים:</strong><br>
• <code>ensure_ascii=False</code> - בלעדיו, עברית תישמר כ-\u05D3... (בלתי קריא!)<br>
• <code>indent=2</code> - מוסיף הזחות ושורות חדשות, כך שהקובץ קריא ויפה.<br>
<strong>בלי indent:</strong> הכל שורה אחת ארוכה. <strong>עם indent:</strong> מעוצב ונוח לקריאה!
</p>

### <p style="align: right; direction: rtl; float: right;">המרה בין Python ל-JSON</p>

<p style="text-align: right; direction: rtl;">
לפעמים נרצה להמיר JSON <strong>למחרוזת</strong> (ולא לקובץ) - למשל כשעובדים עם APIs.<br>
בשביל זה יש <code>json.dumps()</code> (dump <strong>s</strong>tring) ו-<code>json.loads()</code> (load <strong>s</strong>tring).
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה שימושי?</strong><br>
• <strong>APIs</strong> - כששולחים נתונים לשרת או מקבלים ממנו, זה בדרך כלל מחרוזת JSON<br>
• <strong>העברת נתונים</strong> - לפעמים צריך להעביר JSON כטקסט (למשל, בהודעה)<br>
• <strong>ניפוי באגים</strong> - לראות בדיוק איך הנתונים נראים כ-JSON
</p>

```python
# Python dict -> JSON string (המרה למחרוזת)
data = {"name": "שרה", "age": 30, "city": "תל אביב"}
json_string = json.dumps(data, ensure_ascii=False)

print("Python dict:")
print(data)
print(f"Type: {type(data)}\n")

print(f"JSON string:")
print(json_string)
print(f"Type: {type(json_string)}\n")

# JSON string -> Python dict (המרה חזרה)
back_to_python = json.loads(json_string)
print(f"חזרה ל-Python:")
print(back_to_python)
print(f"Type: {type(back_to_python)}")
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
<code>json.dumps(data)</code> המיר את המילון <strong>למחרוזת</strong> JSON - שימו לב שהסוג הוא <code>str</code>!<br>
<code>json.loads(json_string)</code> המיר את המחרוזת <strong>חזרה למילון</strong> פייתון.<br>
<strong>זכרו את ההבדל:</strong><br>
• <code>dump</code> / <code>load</code> - עובדים עם <strong>קבצים</strong><br>
• <code>dumps</code> / <code>loads</code> - עובדים עם <strong>מחרוזות</strong> (ה-s = string)<br>
כשעובדים עם APIs, תמיד תשתמשו ב-<code>dumps</code> ו-<code>loads</code>!
</p>

## <p style="align: right; direction: rtl; float: right;">חלק 4: עדכון קבצי JSON</p>

<p style="text-align: right; direction: rtl;">
לעתים קרובות נרצה <strong>לעדכן</strong> קובץ JSON קיים - לא למחוק הכל ולכתוב מחדש, אלא לשנות רק חלק מהנתונים.
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה חשוב?</strong><br>
דמיינו אפליקציה שמעקבת אחרי פעילות משתמש.<br>
כל פעם שהמשתמש נכנס, צריך לעדכן: מספר כניסות +1, תאריך כניסה אחרון.<br>
לא רוצים למחוק את כל הפרופיל - רק לעדכן את מה שצריך!
</p>

```python
# קריאת ההגדרות הנוכחיות
with open('user_profile.json', 'r', encoding='utf-8') as file:
    profile = json.load(file)

print("📝 עדכון הגדרות משתמש:\n")

# עדכון ערכים - פשוט משנים את המילון!
profile['stats']['login_count'] += 1  # כניסה נוספת
profile['stats']['last_login'] = "2024-01-16"  # תאריך חדש
profile['interests'].append("צילום")  # תחום עניין חדש
profile['preferences']['theme'] = "light"  # החלפת ערכת נושא

print(f"מספר כניסות עודכן ל-{profile['stats']['login_count']}")
print(f"תחומי עניין: {', '.join(profile['interests'])}")
print(f"ערכת נושא: {profile['preferences']['theme']}")

# שמירה חזרה לקובץ - דורסים את הקובץ הישן
with open('user_profile.json', 'w', encoding='utf-8') as file:
    json.dump(profile, file, ensure_ascii=False, indent=2)

print("\n💾 השינויים נשמרו")
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
התהליך הוא <strong>קרא → עדכן → שמור</strong>:<br>
1. קראנו את הקובץ למילון פייתון<br>
2. עדכנו ערכים במילון (בדיוק כמו עבודה עם מילון רגיל!)<br>
3. שמרנו את המילון המעודכן בחזרה לקובץ<br>
<strong>שימו לב:</strong> <code>profile['interests'].append("צילום")</code> - מוסיפים לרשימה שבתוך המילון!<br>
ו-<code>profile['stats']['login_count'] += 1</code> - מגדילים ערך מספרי שמקונן ברמה השנייה.<br>
JSON מאפשר לעבוד עם מבנים מורכבים בקלות!
</p>

## <p style="align: right; direction: rtl; float: right;">חלק 5: המרה בין CSV ל-JSON</p>

<p style="text-align: right; direction: rtl;">
לעתים קרובות נצטרך <strong>להמיר</strong> בין פורמטים.<br>
למשל: קיבלנו נתונים ב-CSV מהבנק, אבל ה-API שלנו עובד רק עם JSON.
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה שימושי?</strong><br>
• <strong>CSV → JSON</strong>: כשרוצים להעביר נתוני טבלה ל-API או לאפליקציה<br>
• <strong>JSON → CSV</strong>: כשרוצים לפתוח נתונים מ-API ב-Excel לניתוח<br>
כל פורמט טוב למשהו אחר - חשוב לדעת להמיר ביניהם!
</p>

```python
import csv

# CSV → JSON: קריאת CSV ושמירה כ-JSON
employees = []
with open('../../resources/employees.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)  # כל שורה = מילון
    for row in reader:
        employees.append(row)  # הוספה לרשימה

# שמירה כ-JSON
with open('employees.json', 'w', encoding='utf-8') as file:
    json.dump(employees, file, ensure_ascii=False, indent=2)

print(f"✅ {len(employees)} עובדים הומרו מ-CSV ל-JSON\n")

# הצגת עובד אחד כדוגמה
print("דוגמה - עובד ראשון ב-JSON:")
print(json.dumps(employees[0], ensure_ascii=False, indent=2))
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
קראנו את ה-CSV עם <code>DictReader</code> - כל שורה הפכה למילון.<br>
אספנו את כל המילונים לרשימה: <code>[{...}, {...}, {...}]</code><br>
ושמרנו את הרשימה כולה כ-JSON!<br>
<strong>התוצאה:</strong> במקום קובץ CSV עם שורות, יש לנו JSON עם מערך של אובייקטים.<br>
זה הפורמט המועדף לעבודה עם APIs ואפליקציות מודרניות!
</p>

### <p style="align: right; direction: rtl; float: right;">המרה מ-JSON חזרה ל-CSV</p>

<p style="text-align: right; direction: rtl;">
כיוון ההפוך - לקחת JSON ולהפוך אותו ל-CSV.<br>
זה שימושי כשקיבלנו נתונים מ-API ורוצים לנתח אותם ב-Excel.
</p>

```python
# JSON → CSV: קריאת JSON ושמירה כ-CSV
with open('employees.json', 'r', encoding='utf-8') as file:
    data = json.load(file)  # רשימה של מילונים

# המרה ל-CSV
with open('employees_from_json.csv', 'w', encoding='utf-8', newline='') as file:
    if data:  # אם יש נתונים
        fieldnames = data[0].keys()  # שמות העמודות מהמילון הראשון
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        
        writer.writeheader()  # כתיבת כותרות
        for row in data:
            writer.writerow(row)  # כתיבת כל שורה

print("✅ JSON הומר בחזרה ל-CSV")
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
קראנו את ה-JSON - קיבלנו רשימה של מילונים (כמו שהיה לנו קודם).<br>
השתמשנו ב-<code>data[0].keys()</code> כדי לקבל את שמות העמודות מהמילון הראשון.<br>
ואז כתבנו כל מילון כשורה ב-CSV עם <code>DictWriter</code>.<br>
<strong>התוצאה:</strong> קובץ CSV זהה למה שהיה לנו בהתחלה!<br>
המעגל נסגר: CSV → JSON → CSV. יכולים להמיר בחופשיות בין הפורמטים!
</p>

JSON הוא הפורמט המושלם לזה - גמיש, קריא, וקל לעדכן.

```python
class ConfigManager:
    """מנהל הגדרות מתקדם למערכת"""
    
    def __init__(self, config_file='config.json'):
        self.config_file = config_file
        self.config = self._load_config()
    
    def _load_config(self):
        """טעינת הגדרות מהקובץ"""
        try:
            with open(self.config_file, 'r', encoding='utf-8') as file:
                return json.load(file)
        except FileNotFoundError:
            # אם הקובץ לא קיים, מחזירים הגדרות ברירת מחדל
            return {
                "app_name": "My App",
                "version": "1.0.0",
                "debug": False,
                "features": {},
                "limits": {}
            }
    
    def _save_config(self):
        """שמירת הגדרות לקובץ"""
        with open(self.config_file, 'w', encoding='utf-8') as file:
            json.dump(self.config, file, ensure_ascii=False, indent=2)
    
    def get(self, key, default=None):
        """קבלת ערך הגדרה - תומך בנתיבים מקוננים!"""
        keys = key.split('.')  # תמיכה ב-'email.smtp_server'
        value = self.config
        
        # ניווט דרך הרמות
        for k in keys:
            if isinstance(value, dict) and k in value:
                value = value[k]  # ירידה רמה למטה
            else:
                return default  # אם המפתח לא קיים, מחזירים ברירת מחדל
        
        return value
    
    def set(self, key, value):
        """הגדרת ערך - תומך ביצירת מבנים מקוננים!"""
        keys = key.split('.')
        current = self.config
        
        # ניווט עד לרמה האחרונה ויצירת מבנה אם צריך
        for k in keys[:-1]:
            if k not in current:
                current[k] = {}  # יצירת מילון חדש אם לא קיים
            current = current[k]
        
        # הגדרת הערך ברמה האחרונה
        current[keys[-1]] = value
        self._save_config()
        
        return f"✅ {key} = {value}"
    
    def delete(self, key):
        """מחיקת הגדרה"""
        keys = key.split('.')
        current = self.config
        
        # ניווט עד לרמה האחרונה
        for k in keys[:-1]:
            if k not in current:
                return f"❌ {key} לא נמצא"
            current = current[k]
        
        # מחיקה
        if keys[-1] in current:
            del current[keys[-1]]
            self._save_config()
            return f"🗑️  {key} נמחק"
        else:
            return f"❌ {key} לא נמצא"
    
    def show_all(self):
        """הצגת כל ההגדרות"""
        print("\n⚙️  הגדרות המערכת:")
        print("=" * 50)
        print(json.dumps(self.config, ensure_ascii=False, indent=2))


# נסו את המערכת!
config = ConfigManager()

# הגדרת ערכים - שימו לב לשימוש בנקודות למבנים מקוננים!
print(config.set("app_name", "מערכת ניהול משימות"))
print(config.set("debug", True))
print(config.set("features.dark_mode", True))  # יוצר features.dark_mode אוטומטית!
print(config.set("features.auto_save", True))
print(config.set("limits.max_users", 100))
print(config.set("limits.max_file_size_mb", 50))

# קריאת ערכים
print(f"\n📖 קריאת ערכים:")
print(f"  שם אפליקציה: {config.get('app_name')}")
print(f"  מצב debug: {config.get('debug')}")
print(f"  מצב לילה: {config.get('features.dark_mode')}")
print(f"  מקסימום משתמשים: {config.get('limits.max_users')}")
print(f"  ערך לא קיים: {config.get('missing.key', 'ברירת מחדל')}")

# הצגת הכל
config.show_all()
```

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
בנינו מערכת שלמה לניהול הגדרות! בואו נבין את הרעיונות המרכזיים:<br>
• <strong>נתיבים מקוננים:</strong> <code>config.get('features.dark_mode')</code> - במקום <code>config['features']['dark_mode']</code><br>
• <strong>ברירת מחדל:</strong> אם הקובץ לא קיים, מתחילים עם הגדרות בסיסיות<br>
• <strong>שמירה אוטומטית:</strong> כל <code>set()</code> או <code>delete()</code> שומרים ישירות לקובץ<br>
• <strong>יצירה דינמית:</strong> <code>set('new.nested.key', value)</code> יוצר את כל המבנה אם לא קיים!<br>
<strong>למה זה מעולה?</strong> זה בדיוק איך מערכות הגדרות עובדות בעולם האמיתי!<br>
במקום לעבוד ישירות עם JSON, יש לנו API נוח שמטפל בהכל.
