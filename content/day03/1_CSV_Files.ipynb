<p style="text-align: right; direction: rtl;">
היום נלמד לעבוד עם קבצי CSV - פורמט נפוץ מאוד לייבוא וייצוא נתונים.<br>

## <p style="align: right; direction: rtl; float: right;">חלק 1: מה זה CSV?</p>

<p style="text-align: right; direction: rtl;">
<strong>CSV</strong> = Comma Separated Values (ערכים מופרדים בפסיקים)<br>
זה קובץ טקסט רגיל שבו כל שורה מייצגת שורה בטבלה, והערכים מופרדים בפסיקים.
</p>

<p style="text-align: right; direction: rtl;">
<strong>דוגמה:</strong>
</p>

```
name,age,city
דני,25,תל אביב
שרה,30,חיפה
יוסי,28,ירושלים
```

<p style="text-align: right; direction: rtl;">
<strong>למה CSV כל כך שימושי?</strong>
</p>

<ul style="text-align: right; direction: rtl;">
    <li><strong>תאימות מושלמת עם Excel</strong> - אפשר לפתוח ולערוך ב-Excel, Google Sheets ועוד</li>
    <li><strong>קל לייצא ממערכות אחרות</strong> - כל מערכת יודעת לייצא CSV (בנקים, CRM, מערכות חשבונאות)</li>
    <li><strong>קובץ טקסט פשוט</strong> - קל לקרוא, לערוך ולתקן גם עם Notepad</li>
    <li><strong>יעיל במיוחד</strong> - עובד מהר גם עם מיליוני שורות (הרבה יותר מהיר מ-Excel!)</li>
    <li><strong>אוניברסלי</strong> - עובד על כל מערכת הפעלה, בכל שפת תכנות</li>
</ul>

<p style="text-align: right; direction: rtl;">
<strong>מתי נשתמש ב-CSV?</strong><br>
כשאנחנו צריכים לעבוד עם נתוני טבלאות, לייבא נתונים ממערכות אחרות, לנתח כמויות גדולות של מידע, או פשוט כשאנחנו רוצים לשמור נתונים בצורה פשוטה ויעילה.
</p>

## <p style="align: right; direction: rtl; float: right;">חלק 2: קריאת קבצי CSV</p>

### <p style="align: right; direction: rtl; float: right;">DictReader - הדרך הנוחה</p>

<p style="text-align: right; direction: rtl;">
פייתון מציעה לנו שתי דרכים לקרוא CSV: <code>reader</code> ו-<code>DictReader</code>.<br>
<code>DictReader</code> היא הדרך הנוחה והמומלצת!
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה DictReader עדיף?</strong><br>
במקום לגשת לעמודות לפי אינדקס (<code>row[0]</code>, <code>row[1]</code>...) שקשה לזכור מה כל עמודה מייצגת,<br>
<code>DictReader</code> הופך כל שורה למילון שבו המפתחות הם <strong>שמות העמודות</strong>!<br>
זה אומר שנוכל לכתוב <code>row['first_name']</code> במקום <code>row[0]</code> - הרבה יותר ברור!
</p>

````python
import csv

# קריאת קובץ CSV עם DictReader
with open('../../resources/employees.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)  # DictReader קורא את השורה הראשונה ככותרות!
    
    print("📋 רשימת עובדים:\n")
    for row in reader:
        # כל שורה היא מילון! המפתחות = שמות העמודות מהשורה הראשונה
        print(f"👤 {row['first_name']} {row['last_name']}")
        print(f"   מחלקה: {row['department']}")
        print(f"   שכר: ₪{row['salary']}")
        print()
````

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
<code>DictReader</code> קרא את השורה הראשונה בקובץ (<code>first_name,last_name,department,salary</code>) והפך אותה לשמות המפתחות במילון.<br>
אחר כך, כל שורה נוספת בקובץ הפכה למילון שבו כל ערך משוייך לשם העמודה המתאים.<br>
זה חוסך לנו את הצורך לזכור "עמודה 0 זה שם פרטי, עמודה 1 זה שם משפחה..." - פשוט כותבים את שם העמודה!
</p>

### <p style="align: right; direction: rtl; float: right;">שמירה לרשימה</p>

<p style="text-align: right; direction: rtl;">
לפעמים נרצה לטעון את <strong>כל</strong> הנתונים מה-CSV לזיכרון כדי לעבוד איתם.<br>
זה שימושי כשאנחנו צריכים לעבור על הנתונים יותר מפעם אחת, או כשאנחנו רוצים לבצע חישובים על כל הנתונים.
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה חשוב?</strong><br>
ברגע שהנתונים ברשימה, אפשר להשתמש בכל כלי העבודה החזקים של פייתון: <code>max()</code>, <code>min()</code>, <code>sum()</code>, list comprehensions ועוד.
</p>

````python
# קריאה ושמירה לרשימה
employees = []

with open('../../resources/employees.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        employees.append(row)  # הוספת כל שורה (מילון) לרשימה

print(f"טענו {len(employees)} עובדים\n")

# עכשיו אפשר לעבוד עם הנתונים!
# מציאת העובד עם השכר הגבוה ביותר
highest_paid = max(employees, key=lambda e: int(e['salary']))
print(f"העובד עם השכר הגבוה ביותר:")
print(f"  {highest_paid['first_name']} {highest_paid['last_name']} - ₪{highest_paid['salary']}")

# חישוב שכר ממוצע
avg_salary = sum(int(e['salary']) for e in employees) / len(employees)
print(f"\nשכר ממוצע: ₪{avg_salary:,.2f}")
````

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
קראנו את כל השורות מהקובץ ושמרנו אותן ברשימה בשם <code>employees</code>.<br>
עכשיו כל עובד הוא מילון ברשימה, ואנחנו יכולים להשתמש ב-<code>max()</code> כדי למצוא את העובד עם השכר הגבוה ביותר,<br>
וב-<code>sum()</code> + generator expression כדי לחשב את השכר הממוצע של כולם.<br>
שימו לב ש-<code>int(e['salary'])</code> הופך את המחרוזת למספר - CSV תמיד קורא הכל כמחרוזות!
</p>

### <p style="align: right; direction: rtl; float: right;">סינון וחיפוש</p>

<p style="text-align: right; direction: rtl;">
אחד היתרונות הגדולים של טעינת הנתונים לרשימה הוא היכולת <strong>לסנן</strong> אותם בקלות.<br>
נוכל למצוא עובדים לפי מחלקה, שכר, או כל קריטריון אחר שנבחר.
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה שימושי בעולם האמיתי?</strong><br>
דמיינו שאתם מנהלי משאבי אנוש וצריכים למצוא את כל העובדים במחלקה מסוימת,<br>
או את כל מי שמעל שכר מסוים לקידום. סינון הוא הכלי שלכם!
</p>

````python
# מציאת כל העובדים במחלקת הנדסה
engineers = [e for e in employees if e['department'] == 'הנדסה']

print(f"🔧 עובדי הנדסה ({len(engineers)} עובדים):\n")
for eng in engineers:
    print(f"  • {eng['first_name']} {eng['last_name']} - ₪{eng['salary']}")

# עובדים עם שכר מעל 14,000
high_earners = [e for e in employees if int(e['salary']) > 14000]

print(f"\n💰 עובדים עם שכר מעל ₪14,000 ({len(high_earners)} עובדים):\n")
for emp in high_earners:
    print(f"  • {emp['first_name']} {emp['last_name']} - {emp['department']} - ₪{emp['salary']}")
````

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
השתמשנו ב-<strong>list comprehension</strong> עם תנאי (<code>if</code>) כדי לסנן את הרשימה.<br>
בשורה הראשונה, עברנו על כל העובדים ושמרנו רק את אלה שהמחלקה שלהם היא "הנדסה".<br>
בשורה השנייה, שמרנו רק עובדים שהשכר שלהם (אחרי המרה למספר!) גדול מ-14,000.<br>
זה מאוד יעיל ונוח - שורה אחת של קוד במקום לולאה עם <code>append</code>!
</p>

## <p style="align: right; direction: rtl; float: right;">חלק 3: כתיבה לקבצי CSV</p>

### <p style="align: right; direction: rtl; float: right;">DictWriter - כתיבה פשוטה</p>

<p style="text-align: right; direction: rtl;">
כמו שיש <code>DictReader</code> לקריאה, יש גם <code>DictWriter</code> לכתיבה!<br>
זה מאפשר לנו לכתוב מילונים ישירות ל-CSV, בלי להתעסק עם סדר העמודות.
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה חשוב?</strong><br>
כשיש לנו נתונים במבנה של מילונים (מאוד נפוץ בפייתון!), <code>DictWriter</code> עושה את העבודה הכי קלה:<br>
פשוט מגדירים את שמות העמודות פעם אחת, ואז כותבים מילונים בלי לדאוג לסדר הנכון.
</p>

````python
# יצירת רשימת סטודנטים - דוגמה מעשית!
students = [
    {'name': 'דני כהן', 'id': 12345, 'grade': 85, 'course': 'מתמטיקה'},
    {'name': 'שרה לוי', 'id': 12346, 'grade': 92, 'course': 'פיזיקה'},
    {'name': 'יוסי מזרחי', 'id': 12347, 'grade': 78, 'course': 'כימיה'},
    {'name': 'רחל אברהם', 'id': 12348, 'grade': 95, 'course': 'מתמטיקה'},
]

# כתיבה לקובץ CSV
with open('students.csv', 'w', encoding='utf-8', newline='') as file:
    # הגדרת השדות (כותרות) - הסדר חשוב!
    fieldnames = ['name', 'id', 'grade', 'course']
    
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    # כתיבת שורת הכותרות (name,id,grade,course)
    writer.writeheader()
    
    # כתיבת כל השורות - פשוט מעביר כל מילון!
    for student in students:
        writer.writerow(student)

print("✅ הקובץ students.csv נוצר בהצלחה!")

# קריאה חזרה לאימות
print("\nתוכן הקובץ:")
with open('students.csv', 'r', encoding='utf-8') as file:
    print(file.read())
````

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
יצרנו רשימה של 4 סטודנטים, כל אחד מיוצג כמילון.<br>
<code>DictWriter</code> קיבל את רשימת השמות של העמודות (<code>fieldnames</code>), וזה קבע את סדר העמודות בקובץ.<br>
<code>writeheader()</code> כתב את שורת הכותרות, ואז <code>writerow()</code> כתב כל סטודנט כשורה חדשה.<br>
<strong>שימו לב</strong> ל-<code>newline=''</code> - זה חשוב! בלעדיו יכולות להיווצר שורות ריקות מיותרות בקובץ.<br>
ואת <code>encoding='utf-8'</code> - חובה לעברית!
</p>

### <p style="align: right; direction: rtl; float: right;">הוספה לקובץ קיים</p>

<p style="text-align: right; direction: rtl;">
לפעמים לא רוצים למחוק קובץ קיים אלא <strong>להוסיף</strong> לו שורות חדשות.<br>
בדיוק בשביל זה קיים מצב <code>'a'</code> (append) במקום <code>'w'</code> (write).
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה שימושי?</strong><br>
דמיינו שיש לכם קובץ עם נתוני מכירות, ובסוף כל יום אתם רוצים להוסיף את המכירות של היום.<br>
במקום לטעון הכל, להוסיף ולשמור מחדש - פשוט פותחים במצב append ומוסיפים!
</p>

````python
# הוספת סטודנטים חדשים לקובץ קיים
new_students = [
    {'name': 'משה דוד', 'id': 12349, 'grade': 88, 'course': 'ביולוגיה'},
    {'name': 'מיכל שמעון', 'id': 12350, 'grade': 91, 'course': 'פיזיקה'},
]

# פתיחה במצב append (הוספה) - שימו לב ל-'a' במקום 'w'!
with open('students.csv', 'a', encoding='utf-8', newline='') as file:
    fieldnames = ['name', 'id', 'grade', 'course']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    # לא כותבים header כי הוא כבר קיים!
    for student in new_students:
        writer.writerow(student)

print("✅ סטודנטים חדשים נוספו!\n")

# קריאה והצגה של כל הסטודנטים
with open('students.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    print("📚 כל הסטודנטים:\n")
    for student in reader:
        print(f"  {student['name']} - {student['course']} - {student['grade']}")
````

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
פתחנו את הקובץ במצב <code>'a'</code> (append) במקום <code>'w'</code> (write).<br>
זה אומר שהתוכן הקיים נשמר, והשורות החדשות נוספות בסוף הקובץ.<br>
<strong>חשוב מאוד:</strong> לא קראנו <code>writeheader()</code> הפעם! למה? כי שורת הכותרות כבר קיימת מהפעם הקודמת.<br>
אם היינו כותבים <code>writeheader()</code> שוב, היינו מקבלים שורת כותרות כפולה באמצע הקובץ.
</p>

## <p style="align: right; direction: rtl; float: right;">חלק 4: עיבוד ומניפולציה</p>

### <p style="align: right; direction: rtl; float: right;">חישוב סטטיסטיקות</p>

<p style="text-align: right; direction: rtl;">
אחד השימושים החשובים ביותר ב-CSV הוא <strong>ניתוח נתונים</strong>.<br>
נניח שיש לנו קובץ CSV עם נתוני מכירות מהחודש האחרון - איך נוציא מזה תובנות?
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה חשוב?</strong><br>
בעולם האמיתי, הנתונים שלנו לא שווים הרבה אם אנחנו לא יודעים לנתח אותם.<br>
מנהל עסק רוצה לדעת: מה המוצר הכי נמכר? מה ההכנסה הממוצעת? איזו קטגוריה מניבה הכי הרבה כסף?<br>
כל התשובות האלה מגיעות מחישובים פשוטים על נתוני CSV.
</p>

````python
# קריאת קובץ המכירות
sales = []
with open('../../resources/sales.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        sales.append(row)

print("📊 סטטיסטיקות מכירות:\n")

# סך המכירות - סכום של כל עמודת 'total'
total_sales = sum(float(sale['total']) for sale in sales)
print(f"סך כל המכירות: ₪{total_sales:,.2f}")

# כמות פריטים שנמכרו - סכום של כל עמודת 'quantity'
total_items = sum(int(sale['quantity']) for sale in sales)
print(f"סך פריטים שנמכרו: {total_items:,}")

# מכירה ממוצעת
avg_sale = total_sales / len(sales)
print(f"מכירה ממוצעת: ₪{avg_sale:,.2f}")

# המוצר היקר ביותר שנמכר
most_expensive = max(sales, key=lambda s: float(s['price']))
print(f"\nהמוצר היקר ביותר: {most_expensive['product']} - ₪{most_expensive['price']}")

# המכירה הגדולה ביותר (מחיר כפול כמות)
biggest_sale = max(sales, key=lambda s: float(s['total']))
print(f"המכירה הגדולה ביותר: {biggest_sale['product']} (כמות: {biggest_sale['quantity']}) - ₪{biggest_sale['total']}")
````

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
השתמשנו בפונקציות מובנות של פייתון כדי לחשב סטטיסטיקות:<br>
• <code>sum()</code> עם generator expression - חישבה את סכום כל ההכנסות וכל הכמויות<br>
• <code>max()</code> עם <code>key</code> - מצאה את המוצר עם המחיר הגבוה ביותר ואת המכירה הגדולה ביותר<br>
• <code>float()</code> ו-<code>int()</code> - המרנו מחרוזות למספרים כדי שנוכל לעבוד איתם מתמטית<br>
<strong>שימו לב:</strong> CSV תמיד קורא הכל כמחרוזות! אם רוצים לעשות חישובים, חייבים להמיר למספרים.
</p>

### <p style="align: right; direction: rtl; float: right;">קיבוץ לפי קטגוריה</p>

<p style="text-align: right; direction: rtl;">
לפעמים נרצה לקבץ נתונים לפי ערך מסוים - למשל, כל המכירות לפי קטגוריית מוצר.<br>
זה מאפשר לנו לענות על שאלות כמו "איזו קטגוריה מכרה הכי הרבה?" או "כמה הכנסה הייתה מאלקטרוניקה?"
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה שימושי?</strong><br>
בניתוח עסקי, קיבוץ הוא אחד הכלים החשובים ביותר.<br>
זה עוזר לראות תמונה רחבה: אילו קטגוריות עובדות טוב? אילו לקוחות קונים הכי הרבה? אילו חודשים הכי רווחיים?
</p>

````python
from collections import defaultdict

# קיבוץ מכירות לפי קטגוריה
# defaultdict יוצר אוטומטית רשימה ריקה לכל מפתח חדש
sales_by_category = defaultdict(list)

for sale in sales:
    category = sale['category']
    sales_by_category[category].append(sale)  # הוספת המכירה לקטגוריה המתאימה

print("📦 מכירות לפי קטגוריה:\n")

for category, category_sales in sales_by_category.items():
    # חישוב סך ההכנסות לכל קטגוריה
    total = sum(float(s['total']) for s in category_sales)
    count = len(category_sales)
    print(f"{category}:")
    print(f"  מספר מכירות: {count}")
    print(f"  סך הכנסות: ₪{total:,.2f}")
    print()
````

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
השתמשנו ב-<code>defaultdict(list)</code> - מבנה נתונים מיוחד שיוצר אוטומטית רשימה ריקה לכל מפתח חדש.<br>
עברנו על כל המכירות, וכל מכירה הוספנו לרשימה של הקטגוריה שלה.<br>
בסוף, עברנו על כל קטגוריה וחישבנו כמה מכירות היו בה וסך ההכנסות ממנה.<br>
<strong>למה defaultdict?</strong> בלעדיו היינו צריכים לבדוק אם הקטגוריה קיימת לפני כל הוספה - <code>defaultdict</code> עושה את זה אוטומטית!
</p>

### <p style="align: right; direction: rtl; float: right;">יצירת דוח מסוכם</p>

<p style="text-align: right; direction: rtl;">
לעתים קרובות נרצה לקחת נתוני גלם ולהפוך אותם ל<strong>דוח מסכם</strong> שקל לקרוא ולהבין.<br>
במקום אלפי שורות של מכירות, נרצה סיכום קצר: כמה הכנסות לכל קטגוריה, מה הממוצע, וכו'.
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה חשוב?</strong><br>
מנהלים לא רוצים לראות אלפי שורות של נתונים - הם רוצים <strong>תובנות</strong>.<br>
דוח מסוכם הופך נתונים גולמיים למידע שימושי שאפשר לקבל על בסיסו החלטות.
</p>

````python
# יצירת דוח מסכם לפי קטגוריה
summary = []

for category, category_sales in sales_by_category.items():
    total = sum(float(s['total']) for s in category_sales)
    quantity = sum(int(s['quantity']) for s in category_sales)
    avg = total / len(category_sales)  # הכנסה ממוצעת למכירה
    
    # בניית מילון עם כל הסטטיסטיקות
    summary.append({
        'category': category,
        'total_sales': total,
        'total_quantity': quantity,
        'num_transactions': len(category_sales),
        'average_sale': avg
    })

# מיון לפי סך הכנסות (מהגבוה לנמוך)
summary.sort(key=lambda x: x['total_sales'], reverse=True)

# שמירת הדוח לקובץ CSV חדש
with open('sales_summary.csv', 'w', encoding='utf-8', newline='') as file:
    fieldnames = ['category', 'total_sales', 'total_quantity', 'num_transactions', 'average_sale']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    writer.writeheader()
    for row in summary:
        writer.writerow(row)

print("✅ דוח מסכם נשמר ל-sales_summary.csv\n")

# הצגת הדוח
print("📈 דוח מכירות מסכם:\n")
print(f"{'קטגוריה':<15} {'הכנסות':<12} {'כמות':<8} {'עסקאות':<10} {'ממוצע'}")
print("=" * 60)
for row in summary:
    print(f"{row['category']:<15} ₪{row['total_sales']:<11,.2f} {row['total_quantity']:<8} "
          f"{row['num_transactions']:<10} ₪{row['average_sale']:.2f}")
````

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
יצרנו דוח מסכם שהופך נתונים גולמיים לתובנות ברורות!<br>
לכל קטגוריה חישבנו: סך הכנסות, כמות כוללת, מספר עסקאות, והכנסה ממוצעת.<br>
אחר כך מיינו את הדוח לפי הכנסות (מהגבוה לנמוך) כדי שנראה מיד איזו קטגוריה הכי חשובה.<br>
ולבסוף - שמרנו את הדוח המסוכם ל-CSV חדש! עכשיו יש לנו קובץ קטן וברור במקום קובץ ענק.<br>
<strong>זה הכוח של עיבוד נתונים:</strong> הפיכת Big Data ל-Smart Data!

## <p style="align: right; direction: rtl; float: right;">חלק 5: פרויקט מעשי - מערכת ניהול מוצרים</p>

<p style="text-align: right; direction: rtl;">
עכשיו נבנה משהו שלם ומעשי - <strong>מערכת ניהול מלאי</strong> לחנות!<br>
המערכת תאפשר להוסיף מוצרים, לעדכן מלאי (כשקונים או מוכרים), ולמצוא מוצרים במלאי נמוך.
</p>

<p style="text-align: right; direction: rtl;">
<strong>למה זה חשוב?</strong><br>
זה בדיוק סוג המערכות שעובדות בעולם האמיתי!<br>
כל חנות צריכה לנהל מלאי - לדעת מה יש, מה חסר, מה נמכר.<br>
CSV הוא פורמט מצוין לזה: פשוט, מהיר, וניתן לייבא ל-Excel לניתוח נוסף.
</p>

````python
class ProductManager:
    """מערכת לניהול מלאי מוצרים"""
    
    def __init__(self, filename='products.csv'):
        self.filename = filename
        self.products = []
        self._load_products()
    
    def _load_products(self):
        """טעינת מוצרים מהקובץ"""
        try:
            with open(self.filename, 'r', encoding='utf-8') as file:
                reader = csv.DictReader(file)
                self.products = list(reader)
                print(f"✅ נטענו {len(self.products)} מוצרים")
        except FileNotFoundError:
            # אם הקובץ לא קיים, זה בסדר - נתחיל עם רשימה ריקה
            print("⚠️  קובץ לא נמצא, יצירת מלאי ריק")
            self.products = []
    
    def _save_products(self):
        """שמירת מוצרים לקובץ"""
        if not self.products:
            return
        
        with open(self.filename, 'w', encoding='utf-8', newline='') as file:
            fieldnames = ['id', 'name', 'category', 'price', 'stock']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            for product in self.products:
                writer.writerow(product)
        
        print("💾 מוצרים נשמרו")
    
    def add_product(self, name, category, price, stock):
        """הוספת מוצר חדש"""
        # יצירת ID חדש - מוצא את ה-ID הגבוה ביותר ומוסיף 1
        if self.products:
            max_id = max(int(p['id']) for p in self.products)
            new_id = max_id + 1
        else:
            new_id = 1
        
        product = {
            'id': str(new_id),
            'name': name,
            'category': category,
            'price': str(price),
            'stock': str(stock)
        }
        
        self.products.append(product)
        self._save_products()
        
        return f"✅ המוצר '{name}' נוסף (ID: {new_id})"
    
    def find_product(self, product_id):
        """מציאת מוצר לפי ID"""
        for product in self.products:
            if product['id'] == str(product_id):
                return product
        return None
    
    def update_stock(self, product_id, quantity):
        """עדכון כמות במלאי - חיובי להוספה, שלילי למכירה"""
        product = self.find_product(product_id)
        if not product:
            return f"❌ מוצר {product_id} לא נמצא"
        
        old_stock = int(product['stock'])
        new_stock = old_stock + quantity
        
        # בדיקה שלא נרד מתחת ל-0
        if new_stock < 0:
            return f"❌ אין מספיק במלאי (קיים: {old_stock})"
        
        product['stock'] = str(new_stock)
        self._save_products()
        
        action = "נוספו" if quantity > 0 else "נמכרו"
        return f"✅ {abs(quantity)} יחידות {action}. מלאי חדש: {new_stock}"
    
    def get_low_stock(self, threshold=10):
        """מוצרים עם מלאי נמוך - חשוב להזמנה מחדש!"""
        return [p for p in self.products if int(p['stock']) < threshold]
    
    def get_products_by_category(self, category):
        """מוצרים לפי קטגוריה"""
        return [p for p in self.products if p['category'] == category]
    
    def show_all_products(self):
        """הצגת כל המוצרים בפורמט נוח לקריאה"""
        if not self.products:
            print("📪 אין מוצרים במערכת")
            return
        
        print("\n📦 רשימת מוצרים:")
        print("=" * 70)
        print(f"{'ID':<5} {'שם':<20} {'קטגוריה':<15} {'מחיר':<10} {'מלאי'}")
        print("=" * 70)
        
        for p in self.products:
            # סימון מלאי נמוך באדום
            stock_indicator = "⚠️" if int(p['stock']) < 10 else "✅"
            print(f"{p['id']:<5} {p['name']:<20} {p['category']:<15} "
                  f"₪{float(p['price']):<9.2f} {stock_indicator} {p['stock']}")


# נסו את המערכת!
manager = ProductManager()

# הוספת מוצרים
print(manager.add_product("מחשב נייד", "אלקטרוניקה", 3500, 15))
print(manager.add_product("עכבר", "אלקטרוניקה", 50, 5))
print(manager.add_product("שולחן", "ריהוט", 800, 8))
print(manager.add_product("כיסא", "ריהוט", 350, 20))

# הצגת כל המוצרים
manager.show_all_products()

# עדכון מלאי
print("\n" + manager.update_stock(1, -2))  # מכירה של 2 מחשבים
print(manager.update_stock(2, 15))  # הזמנה של 15 עכברים

# מוצרים עם מלאי נמוך
low_stock = manager.get_low_stock(10)
print(f"\n⚠️  מוצרים עם מלאי נמוך ({len(low_stock)}):")
for p in low_stock:
    print(f"  • {p['name']} - נותרו {p['stock']} יחידות")
````

<p style="text-align: right; direction: rtl;">
<strong>מה קרה כאן?</strong><br>
בנינו מערכת שלמה לניהול מלאי! בואו נבין איך היא עובדת:<br>
• <code>__init__</code> - בונה את המערכת וטוען מוצרים קיימים מה-CSV (אם יש)<br>
• <code>add_product</code> - מוסיף מוצר חדש עם ID אוטומטי ושומר ל-CSV<br>
• <code>update_stock</code> - מעדכן מלאי (חיובי = קנייה, שלילי = מכירה) ובודק שלא יורדים מתחת ל-0<br>
• <code>get_low_stock</code> - מוצא מוצרים במלאי נמוך (חשוב להזמנה מחדש!)<br>
• <code>show_all_products</code> - מציג טבלה יפה עם כל המוצרים<br>
<strong>למה זה מעולה?</strong> כל פעולה נשמרת ישירות ל-CSV, אז אפילו אם התוכנית נסגרת - הכל שמור!
