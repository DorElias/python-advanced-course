## <p style="align: right; direction: rtl; float: right;">חלק 1: מה זה CSV?</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
<strong>CSV</strong> = Comma Separated Values (ערכים מופרדים בפסיקים)<br>
זה קובץ טקסט רגיל שבו כל שורה מייצגת שורה בטבלה, והערכים מופרדים בפסיקים.
</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
<strong>דוגמה:</strong>
</p>

```
name,age,city
דני,25,תל אביב
שרה,30,חיפה
יוסי,28,ירושלים
```

<p style="text-align: right; direction: rtl; float: right; clear: both;">
<strong>למה CSV שימושי?</strong>
</p>

<ul style="text-align: right; direction: rtl; float: right; clear: both;">
    <li>ניתן לפתוח ב-Excel</li>
    <li>קל לייצא נתונים ממערכות שונות</li>
    <li>קובץ טקסט פשוט - קל לקרוא ולכתוב</li>
    <li>עובד עם כמויות גדולות של נתונים</li>
</ul>

## <p style="align: right; direction: rtl; float: right;">חלק 2: קריאת קבצי CSV</p>

### <p style="align: right; direction: rtl; float: right;">DictReader - הדרך הנוחה</p>

````python
import csv

# קריאת קובץ CSV עם DictReader
with open('../../resources/employees.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    
    print("📋 רשימת עובדים:\n")
    for row in reader:
        # כל שורה היא מילון!
        print(f"👤 {row['first_name']} {row['last_name']}")
        print(f"   מחלקה: {row['department']}")
        print(f"   שכר: ₪{row['salary']}")
        print()
````

### <p style="align: right; direction: rtl; float: right;">שמירה לרשימה</p>

````python
# קריאה ושמירה לרשימה
employees = []

with open('../../resources/employees.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        employees.append(row)

print(f"טענו {len(employees)} עובדים\n")

# עכשיו אפשר לעבוד עם הנתונים
# מציאת העובד עם השכר הגבוה ביותר
highest_paid = max(employees, key=lambda e: int(e['salary']))
print(f"העובד עם השכר הגבוה ביותר:")
print(f"  {highest_paid['first_name']} {highest_paid['last_name']} - ₪{highest_paid['salary']}")

# חישוב שכר ממוצע
avg_salary = sum(int(e['salary']) for e in employees) / len(employees)
print(f"\nשכר ממוצע: ₪{avg_salary:,.2f}")
````

### <p style="align: right; direction: rtl; float: right;">סינון וחיפוש</p>

````python
# מציאת כל העובדים במחלקת הנדסה
engineers = [e for e in employees if e['department'] == 'הנדסה']

print(f"🔧 עובדי הנדסה ({len(engineers)} עובדים):\n")
for eng in engineers:
    print(f"  • {eng['first_name']} {eng['last_name']} - ₪{eng['salary']}")

# עובדים עם שכר מעל 14,000
high_earners = [e for e in employees if int(e['salary']) > 14000]

print(f"\n💰 עובדים עם שכר מעל ₪14,000 ({len(high_earners)} עובדים):\n")
for emp in high_earners:
    print(f"  • {emp['first_name']} {emp['last_name']} - {emp['department']} - ₪{emp['salary']}")
````

## <p style="align: right; direction: rtl; float: right;">חלק 3: כתיבה לקבצי CSV</p>

### <p style="align: right; direction: rtl; float: right;">DictWriter - כתיבה פשוטה</p>

````python
# יצירת רשימת סטודנטים
students = [
    {'name': 'דני כהן', 'id': 12345, 'grade': 85, 'course': 'מתמטיקה'},
    {'name': 'שרה לוי', 'id': 12346, 'grade': 92, 'course': 'פיזיקה'},
    {'name': 'יוסי מזרחי', 'id': 12347, 'grade': 78, 'course': 'כימיה'},
    {'name': 'רחל אברהם', 'id': 12348, 'grade': 95, 'course': 'מתמטיקה'},
]

# כתיבה לקובץ CSV
with open('students.csv', 'w', encoding='utf-8', newline='') as file:
    # הגדרת השדות (כותרות)
    fieldnames = ['name', 'id', 'grade', 'course']
    
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    # כתיבת שורת הכותרות
    writer.writeheader()
    
    # כתיבת כל השורות
    for student in students:
        writer.writerow(student)

print("✅ הקובץ students.csv נוצר בהצלחה!")

# קריאה חזרה לאימות
print("\nתוכן הקובץ:")
with open('students.csv', 'r', encoding='utf-8') as file:
    print(file.read())
````

### <p style="align: right; direction: rtl; float: right;">הוספה לקובץ קיים</p>

````python
# הוספת סטודנטים חדשים לקובץ קיים
new_students = [
    {'name': 'משה דוד', 'id': 12349, 'grade': 88, 'course': 'ביולוגיה'},
    {'name': 'מיכל שמעון', 'id': 12350, 'grade': 91, 'course': 'פיזיקה'},
]

# פתיחה במצב append (הוספה)
with open('students.csv', 'a', encoding='utf-8', newline='') as file:
    fieldnames = ['name', 'id', 'grade', 'course']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    # לא כותבים header כי הוא כבר קיים!
    for student in new_students:
        writer.writerow(student)

print("✅ סטודנטים חדשים נוספו!\n")

# קריאה והצגה
with open('students.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    print("📚 כל הסטודנטים:\n")
    for student in reader:
        print(f"  {student['name']} - {student['course']} - {student['grade']}")
````

## <p style="align: right; direction: rtl; float: right;">חלק 4: עיבוד ומניפולציה</p>

### <p style="align: right; direction: rtl; float: right;">חישוב סטטיסטיקות</p>

````python
# קריאת קובץ המכירות
sales = []
with open('../../resources/sales.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        sales.append(row)

print("📊 סטטיסטיקות מכירות:\n")

# סך המכירות
total_sales = sum(float(sale['total']) for sale in sales)
print(f"סך כל המכירות: ₪{total_sales:,.2f}")

# כמות פריטים שנמכרו
total_items = sum(int(sale['quantity']) for sale in sales)
print(f"סך פריטים שנמכרו: {total_items:,}")

# מכירה ממוצעת
avg_sale = total_sales / len(sales)
print(f"מכירה ממוצעת: ₪{avg_sale:,.2f}")

# המוצר היקר ביותר שנמכר
most_expensive = max(sales, key=lambda s: float(s['price']))
print(f"\nהמוצר היקר ביותר: {most_expensive['product']} - ₪{most_expensive['price']}")

# המכירה הגדולה ביותר
biggest_sale = max(sales, key=lambda s: float(s['total']))
print(f"המכירה הגדולה ביותר: {biggest_sale['product']} (כמות: {biggest_sale['quantity']}) - ₪{biggest_sale['total']}")
````

### <p style="align: right; direction: rtl; float: right;">קיבוץ לפי קטגוריה</p>

````python
from collections import defaultdict

# קיבוץ מכירות לפי קטגוריה
sales_by_category = defaultdict(list)

for sale in sales:
    category = sale['category']
    sales_by_category[category].append(sale)

print("📦 מכירות לפי קטגוריה:\n")

for category, category_sales in sales_by_category.items():
    total = sum(float(s['total']) for s in category_sales)
    count = len(category_sales)
    print(f"{category}:")
    print(f"  מספר מכירות: {count}")
    print(f"  סך הכנסות: ₪{total:,.2f}")
    print()
````

### <p style="align: right; direction: rtl; float: right;">יצירת דוח מסוכם</p>

````python
# יצירת דוח מסכם לפי קטגוריה
summary = []

for category, category_sales in sales_by_category.items():
    total = sum(float(s['total']) for s in category_sales)
    quantity = sum(int(s['quantity']) for s in category_sales)
    avg = total / len(category_sales)
    
    summary.append({
        'category': category,
        'total_sales': total,
        'total_quantity': quantity,
        'num_transactions': len(category_sales),
        'average_sale': avg
    })

# מיון לפי סך הכנסות
summary.sort(key=lambda x: x['total_sales'], reverse=True)

# שמירת הדוח לקובץ חדש
with open('sales_summary.csv', 'w', encoding='utf-8', newline='') as file:
    fieldnames = ['category', 'total_sales', 'total_quantity', 'num_transactions', 'average_sale']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    writer.writeheader()
    for row in summary:
        writer.writerow(row)

print("✅ דוח מסכם נשמר ל-sales_summary.csv\n")

# הצגת הדוח
print("📈 דוח מכירות מסכם:\n")
print(f"{'קטגוריה':<15} {'הכנסות':<12} {'כמות':<8} {'עסקאות':<10} {'ממוצע'}")
print("=" * 60)
for row in summary:
    print(f"{row['category']:<15} ₪{row['total_sales']:<11,.2f} {row['total_quantity']:<8} "
          f"{row['num_transactions']:<10} ₪{row['average_sale']:.2f}")
````

## <p style="align: right; direction: rtl; float: right;">חלק 5: פרויקט מעשי - מערכת ניהול מוצרים</p>

````python
class ProductManager:
    """מערכת לניהול מלאי מוצרים"""
    
    def __init__(self, filename='products.csv'):
        self.filename = filename
        self.products = []
        self._load_products()
    
    def _load_products(self):
        """טעינת מוצרים מהקובץ"""
        try:
            with open(self.filename, 'r', encoding='utf-8') as file:
                reader = csv.DictReader(file)
                self.products = list(reader)
                print(f"✅ נטענו {len(self.products)} מוצרים")
        except FileNotFoundError:
            print("⚠️  קובץ לא נמצא, יצירת מלאי ריק")
            self.products = []
    
    def _save_products(self):
        """שמירת מוצרים לקובץ"""
        if not self.products:
            return
        
        with open(self.filename, 'w', encoding='utf-8', newline='') as file:
            fieldnames = ['id', 'name', 'category', 'price', 'stock']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            for product in self.products:
                writer.writerow(product)
        
        print("💾 מוצרים נשמרו")
    
    def add_product(self, name, category, price, stock):
        """הוספת מוצר חדש"""
        # יצירת ID חדש
        if self.products:
            max_id = max(int(p['id']) for p in self.products)
            new_id = max_id + 1
        else:
            new_id = 1
        
        product = {
            'id': str(new_id),
            'name': name,
            'category': category,
            'price': str(price),
            'stock': str(stock)
        }
        
        self.products.append(product)
        self._save_products()
        
        return f"✅ המוצר '{name}' נוסף (ID: {new_id})"
    
    def find_product(self, product_id):
        """מציאת מוצר לפי ID"""
        for product in self.products:
            if product['id'] == str(product_id):
                return product
        return None
    
    def update_stock(self, product_id, quantity):
        """עדכון כמות במלאי"""
        product = self.find_product(product_id)
        if not product:
            return f"❌ מוצר {product_id} לא נמצא"
        
        old_stock = int(product['stock'])
        new_stock = old_stock + quantity
        
        if new_stock < 0:
            return f"❌ אין מספיק במלאי (קיים: {old_stock})"
        
        product['stock'] = str(new_stock)
        self._save_products()
        
        action = "נוספו" if quantity > 0 else "נמכרו"
        return f"✅ {abs(quantity)} יחידות {action}. מלאי חדש: {new_stock}"
    
    def get_low_stock(self, threshold=10):
        """מוצרים עם מלאי נמוך"""
        return [p for p in self.products if int(p['stock']) < threshold]
    
    def get_products_by_category(self, category):
        """מוצרים לפי קטגוריה"""
        return [p for p in self.products if p['category'] == category]
    
    def show_all_products(self):
        """הצגת כל המוצרים"""
        if not self.products:
            print("📪 אין מוצרים במערכת")
            return
        
        print("\n📦 רשימת מוצרים:")
        print("=" * 70)
        print(f"{'ID':<5} {'שם':<20} {'קטגוריה':<15} {'מחיר':<10} {'מלאי'}")
        print("=" * 70)
        
        for p in self.products:
            stock_indicator = "⚠️" if int(p['stock']) < 10 else "✅"
            print(f"{p['id']:<5} {p['name']:<20} {p['category']:<15} "
                  f"₪{float(p['price']):<9.2f} {stock_indicator} {p['stock']}")


# נסו את המערכת!
manager = ProductManager()

# הוספת מוצרים
print(manager.add_product("מחשב נייד", "אלקטרוניקה", 3500, 15))
print(manager.add_product("עכבר", "אלקטרוניקה", 50, 5))
print(manager.add_product("שולחן", "ריהוט", 800, 8))
print(manager.add_product("כיסא", "ריהוט", 350, 20))

# הצגת כל המוצרים
manager.show_all_products()

# עדכון מלאי
print("\n" + manager.update_stock(1, -2))  # מכירה
print(manager.update_stock(2, 15))  # הזמנה

# מוצרים עם מלאי נמוך
low_stock = manager.get_low_stock(10)
print(f"\n⚠️  מוצרים עם מלאי נמוך ({len(low_stock)}):")
for p in low_stock:
    print(f"  • {p['name']} - נותרו {p['stock']} יחידות")
````
